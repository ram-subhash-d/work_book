```python
image = cv2.imread("peacock_feather.jpg")
print(type(image))
print(image.shape)
image
```

```python
print(image.min())
print(image.max())
```

```python
# Since the original image is RGB, by defaul cv treats the image is of BGR format
# Hence convertion is needed
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
# To show one channel of the image
image_2 = cv2.cvtColor(cv2.imread("rgb.jpg"), cv2.COLOR_BGR2RGB)
plt.subplot(1,3,1)
plt.imshow(image_2)
plt.subplot(1,3,2)
plt.imshow(image_2[:, :, 0], cmap='gray')
plt.subplot(1,3,3)
plt.imshow(image_2[:, :, 2], cmap='gray')
```

```python
#Image resize
print(image_2.shape)
image_2_resize = cv2.resize(image_2, (200, 200))
print(image_2_resize.shape)
plt.imshow(image_2_resize)
```

```python
#Make your own image
image_3 = np.zeros((50, 50, 3))
plt.subplot(2,3,1)
plt.imshow(image_3)
image_3[:,:,:] = 255
plt.subplot(2,3,2)
plt.imshow(image_3)
image_3[:,:,:] = 0
image_3[:10,:10,0] = 255
plt.subplot(2,3,3)
plt.imshow(image_3)
image_3[:10,40:,1] = 255
plt.subplot(2,3,4)
plt.imshow(image_3)
image_3[20:30,20:30,0] = 255
image_3[20:30,20:30,1] = 255
plt.subplot(2,3,5)
plt.imshow(image_3)
# Image cropping
plt.subplot(2,3,6)
plt.imshow(image_3[:10,:,:])
```

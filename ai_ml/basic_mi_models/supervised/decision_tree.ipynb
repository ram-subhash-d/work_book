# Decision Tree
Decision tree is capable of solving both regression and clasification problems. Predictive model in the tree form that maps items to its target value, starting from the root to the leaf. Each branch in the tree represents a decision. We must move from hetrogenious mix of items to homogenious baskets, using a decision(if) at each tree node. Performs well both for binary and multi class clasification.

## Model
**Entropy** : It is the measure of randomness or unpredictability in the dataset. If data has lots of different values then entropy is high. For example if there are lots of different types of vegetables in a basket for a vegetable clasification problem then entropy is high. As we move down the decision tree the entropy should reduce.

Formula : $\sum_{i=1}^{k}p(valuei)log_{2}(p(value1))$  k - number of classes in the output lable, p(valuei) - fraction of ith class in the dataset.


**Information Gain** : It is a measure of decrease in entropy after dataset is split. From the above diagram Information Gain = E1- E2.

**Gini Impurity(GI)** : 1 - $\sum_{i=1}^m f_{i}^2$ , fi is the fraction of the class label i, m - number of classes.

**Gini Split Index** : GI(s) - p1GI(s1) - p2GI(s2) , s - parent node, s1 & s2 - child nodes, p1 & p2 - split fractions.

**Information Entropy(IE)** : -$\sum_{i=1}^m f_i * log(f_i)$

At each step we have to split data such that the information gain is highest. First calculate the entropy using the above formula for the parent node. Apply a condition and split the data. After spliting the data calculate the entropy of each of the two baskets. Sum the information gains. Repeat the above steps with different conditions. Finally pick the condition that is giving maximum information gain. Build a tree like this till entropy of the nodes become zero or almost zero. Conditions are based on the features. Condition should be same at each level of the tree.

We can use any of the above given metrics. If we are using Gini Impurity, then the GI of the leaf nodes must be zero.

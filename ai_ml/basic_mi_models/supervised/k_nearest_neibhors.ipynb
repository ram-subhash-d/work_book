# K-nearest Neighbhors

## Model
k-NN is a non-parametric supervised learning methods. Its philosophy is that if you want to know soemone, understand their neibhors. For the new sample to be classified, we identify the k nearest trainig samples based on some distance metric. Then based on these k neighbhor labels we predict the class of the new sample by majority voting. As we can see there is no training as such here, we predict the new sample directly from the taining data. k is normal picked by doing a trail and error, as there is no structured way to find the best k. For a smaller value of k, noise will have high impact, for a high value of k it becoms computationally expensive. General practice is to start with k = $\sqrt n$

### Distance Metric

#### Numerical Independent Variables  
* Euclidean : Straight line length from the new sample to training sample points
$$\sqrt\sum_{i=1}^{n} (x^i - xtest)^2$$ 
i - each dimension(independent variables)
* Manhattan Distance : Travel along each dimension and add the lengths
$$\sum_{i=1}^{n}|x^i - xtest|$$

#### Strings Of Equal Length
* Hamming distance : Number of positions in which two strings differ

#### Strings Of Unequal Length
* Do Integer Encoding : Each category is given a number, like lion - 1, dog - 2, cheetah - 3 etc. Here the distance between lion and dog is 1 but the distance beteen lion and cheetah is 2, which is not consistant. if we want them to be consistant we go for One-hot.
* Do One-hot Encoding : The concerned vairiable is converted into multiple binary variables. That is lion will be given [1,0,0], dog [0,1,0] and cheetah [0,0,1], if there are 3 unique values in the varaible.

## Validation
#### Confusion Matrix
Used to comment on the performance of the algorithm. Build this table on the test data results. Make the table with predicted labels on the x direction and true labels on the y direction. Fill the table with the number of occurences of each combinaton. Diagnal values on the table are the once that are predicted correctly, all the other are wrongly predicted. Accuracy is eqaul to the count on the diagnal by total count. When there is class imbalance(more samples from one type of class) in the training set mearsuring the models performance only based on accuracy can be misleading. For example if there are 9 samples from one class and 1 sample from another class in the training set and if the model predicts all of the samples to be of class 1, then the accuracy is 90%. Here even though the accuracy is good the model is not good. 

Hence we have to look at class wise metrics - precision and recall, to access the model.
* Recall: Express how well a model is able to detect a class. Mathematically, we define recall as the number of true positives divided by the number of true positives plus the number of false negatives.
* Precision: How trustable is the prediction if a sample is assigned to that class. Mathematically, precision is the number of true positives divided by the number of true positives plus the number of false positives.


High precision and recall are desired, some times people also work with F-score which is a harmonic mean of precision and recall. F1 = (2 * precision * recall)/(precision +  recall).

## Merits
* Super easy to implement
* Good performance in several non-linear problems

## Demerits
* Have to store the entire training data
* All distances have to be calculated - computationally intensive
* Susceptible to noise
* Might not capture global trends in data
* Poor predictions in sparse data region

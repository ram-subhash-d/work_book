## Model
The model being learnt is assumed to be a linear function of the parameters. \
y = a0 + a1f1 + a2f2 + ... + anfn \
f - features(data input variables), a - parameters(to learn and model), y - output(data output variable) \
We have to build a straight line which is a best fit for all the data points, which is decided by the parameters a0, a1, a2 and so on.

## Training
A regression model(line) does not fit all the data. The vertical distance between the data point and the regression line is called the residual or error.\
ei = yi - y'i = yi - (a0 + a1xi) - for a one feature data.\
The sum of the residuals of all data points should be as low as possible. Residuals is the feedback to the model.
Hence it becomes a optimization problem, the objective is to estimate the parameters of the model(a0, a1) such that the sum of the squared errors/residuals(SSE) (loss function) of all the data points is minimised. \

## Validation
Check to see if the model predicts well for unseen data. Model validation is performed to evaluate a trained model in an unbiased manner. When we start we take a portion of the data for validation called the test data set, the remaining used for training is called the training data set(For example we use a 70-30 split, 70% for training and 30% for validation). If the performance is not good either a different model is to be chosen or the hyper parameters need to be tuned.

On this unseen data we measure the loss function(Mean squared error(MSE) or Root mean squared error(RMSE) or R-squared). Goal is to select the model or hyper parameters with the least loss.
R-squared is corelation coefficient between predictions and original ground truth, lies between 0 and 1. \
 - yi : output from data, yi hat : predictions

##### Spliting techniques :

Hold Out : Most basic split where the data is split into training and validation sets in the of 80:20 or 70:30. 20 or 30 % data is held out hence the name.

K-fold cross validation : We divide the entire data into k equal folds or groups. Each time one fold is used for validation and the remaining for training.This is repeated k times by considering each of the folds as validation sets each time and average performance is taken. This removes the sampling bias.

Leave-one-out cross validation : Similar to K-fold.

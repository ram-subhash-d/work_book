## CUDA toolkit
Compute Unified Device Architecture(CUDA) is Nvidia's parallel computing platform, it is a collection of tools. Below are its key components.
* CUDA compiler (nvcc) : Transforms CUDA code into GPU hardware code. It converts CUDA codes into PTX(parallel thread execution) codes. PTX is a low-level, hardware-agnostic virtual instruction set architecture (ISA) that acts as an intermediate representation (IR) for the code compiled from CUDA. The compiler transforms CUDA code into a executable file.
* CUDA libraries : CUDA is enriched by a suite of libraries, each one serving a specialized purpose. Some well known libraries are cuBLAS(basic linear algebra), cuFFT(fast fourier transforms), cuRAND(random number generation), cuDNN(deep neural network), etc.
* CUDA runtime and driver APIs : These provide a interface for managing devices, memory and program execution on GPUs. For example cudaMalloc() allocates memory on the GPU, cudaMemcpy() copies data between CPU memory and GPU memory. The GPU driver is also reponsible for just-in-time compilation to convert PTX into device(GPU) specific codes.
* Debug/performance analysis tools : It has tools like Nsight systems for system wide performance analysis, Nsight compute for kernel profiling and analysis, CUDA-GDB for debugging CUDA applications on Linux and Mac OS, CUDA-MEMCHECK to detect and diagnose memory errors, ect.
* CUDA samples : A collection of sample codes to demonstrate various CUDA programming techniques and right way to use them.
For each new hardware, new collection of CUDA tools will be released making use of the enhanced hardware.

## Installation of CUDA toolkit
* Windows
    * Install latest visual studio community with C++ packages selected.
    * Download the latest CUDA toolkit from Nvidia for your operating system.
    * Install the toolkit.
    * It detects the GPU present on the system and installs the right version of the drivers and other tools.
    * Install the toolkit after installing visual studio community, this way CUDA will be integrated into visaul studio.
    * After installation we can launch the Nvidia control panel, to see information about the GPU present on the system and the corresponding drivers installed.
* Linux
    * sudo apt install nvidia-cuda-toolkit will detect the local GPU and install the right toolkit.
    * nvcc --version will give the version of the compiler installed.

## Install WSL on windows
* Turn on windows subsystem for linux feature
* Command wsl --update will update the wsl to the latest
* Command wsl --set-default-version 2 will set wsl to the latest version 2
* Command wsl --install -d Ubuntu will install the latest ubuntu on your system
* Give in the desired user name and password during the installation

## Inline Namespaces
In C++ we can have namespace's inside other namespace's, called nested namespaces. We can make the inner namespace as inline, then any symbols defined in the inner namespace will appear to be defined in the outer namespace.
```
namespace A
{
    inline namespace B
    {
        int x;
    }
    
    x = 6;  //no need to put namespace B:: here
}

int main()
{
    A::x = 5;  //apear to be defind in namespace A, no need for A::B::
}
```
C++ standard library uses inline namespaces.
```
namespace std
{
    inline namespace literals
    {
        inline namespace string_literals
        {
            ...
        }
        inline namespace chrono_literals
        {
            ...
        }
    }
}
```
This makes everything accessible from the std:: namespace. We can also use inline namespaces for versioning. We put the current version in inline namespace and all the previous versions in regular namespaces.
```
namespace product
{
    namespace version1
    {
        class refrigerator
        {
            ...
        }
    }
    
    inline namespace version2
    {
        class refrigerator
        {
            ...
        }
    }
    
    refrigerator fridge; //uses version2 as its namespace is inline
}
```

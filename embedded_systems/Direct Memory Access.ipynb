## Direct Memory Access
When using Memory Mapped I/O or Port Mapped I/O, processor is the bus master. Here the processor takes control of the bus to move the data from the peripheral register to the processor's internal register and then move it to the memory. In order to reduce this burden from the processor and free it to do other things, a new controller Direct Memeory Access(DMA) has been introduced. DMA controller acts as a bus master and transfers data between peripheral to memory, memory to peripheral and memory to memory. In a micro controller there will be multiple bus masters. The processor itself has 3 masters(I-bus, D-bus and S-bus), a couple of DMA controllers, USB OTG etc. DMA controllers are connected and can act as master's to both main and peripheral buses. There are many slaves connected to these buses, like all peripherals connected to the peripheral buses(I2C, SPI, ADC etc), all memory interfaces etc. All the masters and slaves are connected on a bus matrix. Processor and DMA can do data transfers simultaneously over seperate bus paths on the bus matrix.

## DMA Programming
The generic steps for using DMA are 1) Select which DMA controller to use 2) Initialize the DMA 3) Trigger the DMA(automatic or manual) 4) Wait for transfer complete(poll) or get callback from DMA (interrupt).

## DMA Programming On Micro Controller
Normally the micro controller comes with a DMA driver. We use this driver to program and control the DMA controller. The driver typically has a structure, which has to be created and initialized. Then a driver API to initialize the DMA has to be called with this structure. This structure is a parameter for most of the driver methods. The driver will have another API to start DMA data transfer which takes start address, end address and the number of bytes to transfer. On completion of the transfer or error, DMA will raise an interrupt to the processor. The driver will have a method to register a callback to be raised on this interrupt. There are 2 types of triggers manual and automatic. In manual case the processor will tell the DMA to start the transfer, like the API to start DMA data transfer discussed above. In automatic case the peripheral(DMA source) will be configured to trig the DMA to do the data transfer when it has data, DMA will raise interrupt to the processor on completion of the transfer. Normally the micro controllers respective peripheral driver will have a API to tell the peripheral that it should start the transfer when the data is available.

A university has n courses to offer. To graduate from that university, a student must complete all those courses. Some courses have prerequisite courses. One can take a course only after completing all of its prerequisites. Dependencies between the courses are described by two arrays a and b of the same size: course a{i} must be taken before course b{i} for all valid indices. Is it possible to complete all the courses without violating constraints?  
Example :- Input : {"n": 4,"a": {1, 1, 3},"b": {0, 2, 1}} -> Output : 1.

This is a directed graph as we have to complete the prerequisites before doing the course. To have a possibility to complete all the courses there should be no cycles in the graph. To find cycle in a directed graph we can use arrival/departure(DFS) or kann's algorithm(BFS).

## Solution 1: Arrival/Departure
```C++
bool is_cycle(vector<vector<int>> &adj_list, int node, vector<bool> &arrival, vector<bool> &departure)
{
    arrival[node] = true;
    
    for(int neigh : adj_list[node])
    {
        if(arrival[neigh] == true && departure[neigh] == false)
        {
            return true;
        }
        
        //Fresh node
        if(arrival[neigh] == false && (is_cycle(adj_list, neigh, arrival, departure) == true))
        {
            return true;
        }
    }
    
    departure[node] = true;
    return false;
}

bool can_be_completed(int n, vector<int> &a, vector<int> &b) 
{    
    //Variable to store the result
    bool directed_cycle = false;
    
    //Build the adjacency list
    vector<vector<int>> adj_list(n);
    for(int i = 0; i < a.size(); i++)
    {
        adj_list[a[i]].push_back(b[i]);
    }
    
    //Initialize - arrival/departure states
    vector<bool> arrival;
    arrival.resize(n, false);
    vector<bool> departure;
    departure.resize(n, false);
    
    for(int i = 0; i < n; i++)
    {
        if(arrival[i] == false)
        {
            directed_cycle = is_cycle(adj_list, i, arrival, departure);
            if(directed_cycle)
            {
                break;
            }
        }
    }
     
    return !directed_cycle;
}
```

## Solution 2: Kann's algorithm

```C++
bool can_be_completed(int n, vector<int> &a, vector<int> &b) 
{
    // Write your code here.
    
    //Resultant varaible to store topological sort order
    vector<int> res;

    //Construct the adjacency list
    vector<vector<int>> adj_list(n);
    for(int i = 0; i < a.size(); i++)
    {
        adj_list[a[i]].push_back(b[i]);
    }
    
    //Calculate the indegree of each node
    vector<int> indegree;
    indegree.resize(n, 0);
    for(int i = 0; i < n; i++)
    {
        for(int neigh : adj_list[i])
        {
            indegree[neigh]++;
        }
    }
    
    //BFS queue and push independent elements into it.
    queue<int> level_q;
    for(int i = 0; i < n; i++)
    {
        if(indegree[i] == 0)
        {
            level_q.push(i);
        }
    }
    
    //Kann's algorithm
    while(!level_q.empty())
    {
        int node = level_q.front();
        level_q.pop();
        res.push_back(node);
        
        for(int neigh : adj_list[node])
        {
            indegree[neigh]--;
            if(indegree[neigh] == 0)
            {
                level_q.push(neigh);
            }
        }
    }
    
    //if the resultant topological sorting has all the nodes then no cycle
    bool no_cycle = res.size() == n;
    
    return no_cycle;
}


The Time and space complexity is O(m * n) for both the solutions.

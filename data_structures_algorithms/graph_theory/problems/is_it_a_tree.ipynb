Given an undirected graph, find out whether it is a tree.
The undirected edges are given by two arrays edge_start and edge_end of equal size. Edges of the given graph connect nodes edge_start{i} and edge_end{i} for all valid i.
Example :- Input : {"node_count": 4,"edge_start": {0, 0, 0},"edge_end": {1, 2, 3}} -> Output : 1.

## Solution : DFS/BFS with parent vector
An undirected graph will be a tree, if there are no cycles in the tree and if there is only one component in the tree. To traverse the graph to check for cycle we can either use BFS or DFS, the below code uses BFS.

```C++
//Method to identify if there is a cycle in the tree
bool is_cycle(vector<vector<int>> &adj_list, queue<int> &level_q, unordered_set<int> &visited, vector<int> &parent)
{
    bool cycle = false;
    
    while(!level_q.empty())
    {
        int node = level_q.front();
        level_q.pop();
        
        //push the neighbours
        vector<int> neighbours = adj_list[node];
        for(int i = 0; i < neighbours.size(); i++)
        //for(int &neigh : neighbours)
        {
            int neigh = neighbours[i];
            //if cycle return true
            if((visited.count(neigh) == 1) && (neigh != parent[node]))
            {
                level_q = {};
                cycle = true;
            }
            
            //If already not visited, update parent, mark as visited and add to the queue
            if(visited.count(neigh) == 0)
            {
                visited.insert(neigh);
                parent[neigh] = node;
                level_q.push(neigh);
            }
        }
    }
    
    //if no cycle is found return false
    return cycle;
}

bool is_it_a_tree(int node_count, vector<int> &edge_start, vector<int> &edge_end) 
{    
    vector<pair<int, int>> multiple_edges;
    
    //Make a adjacency list for easy traversal
    vector<vector<int>> adj_list(node_count);
    for(int i = 0; i < edge_start.size(); i++)
    {
        //if there is a self edge return false
        if(edge_start[i] == edge_end[i])
        {
            return false;
        }
        
        //if there are multiple edges between 2 nodes return false
        if(find(multiple_edges.begin(), multiple_edges.end(), pair<int, int>{min(edge_start[i], edge_end[i]), max(edge_start[i], edge_end[i])}) != multiple_edges.end())
        {
            return false;
        }
        
        adj_list[edge_start[i]].push_back(edge_end[i]);
        adj_list[edge_end[i]].push_back(edge_start[i]);
        
        multiple_edges.push_back({min(edge_start[i], edge_end[i]), max(edge_start[i], edge_end[i])});
    }
    
    //Keep track of the visited nodes, and the parent in the search tree
    unordered_set<int> visited;
    vector<int> parent{};
    parent.reserve(node_count);
    for(int i = 0; i < parent.size(); i++)
    {
        parent[i] = -1;
    }
    
    //Queue for BFS, push the root node
    queue<int> level_q;
    level_q.push(edge_start[0]);
    visited.insert(edge_start[0]);
    
    bool is_tree = !is_cycle(adj_list, level_q, visited, parent);
    
    return is_tree && (visited.size() == node_count);
}
```

Time complexity is O(n+m) as we need to traverse all the nodes and edges. Space complexity is O(n+m) for the adjacency list, all other space used is less than this order.

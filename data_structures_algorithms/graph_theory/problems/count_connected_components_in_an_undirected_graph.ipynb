Given an undirected graph, find its total number of connected components.  
Example :- Input : {"n": 5,"edges": {{0 ,1}, {1, 2}, {0, 2}, {3, 4}}} -> Output : 1.

## Solution : DFS or BFS
We can do a DFS or a BFS to traverse through all the nodes in a component, then we can check for unvisited nodes to find other components. Time and space complexity is same for both DFS and BFS. This solution uses DFS.

```C++
void depth_first_search(vector<vector<int>> &adjacency_list, int node, unordered_set<int> &visited)
{
    if(visited.count(node) == 1)
    {
        return;
    }
    
    visited.insert(node);
    for(int n_node : adjacency_list[node])
    {
        if(visited.count(n_node) == 0)
        {
            depth_first_search(adjacency_list, n_node, visited);
        }
    }
}

int number_of_connected_components(int n, vector<vector<int>> &edges) 
{   
    //Convert the edges into adjacency list
    vector<vector<int>> adjacency_list(n);
    for(vector<int> &edge : edges)
    {
        adjacency_list[edge[0]].push_back(edge[1]);
        adjacency_list[edge[1]].push_back(edge[0]);
    }
    
    //Set to keep track of all the visited nodes
    unordered_set<int> visited;
    
    //Perform DFS and count all the components
    int components_count{0};
    for(int i = 0; i < n; i++)
    {
        //If node not visited increament count and do a dfs to traverse the full component
        if(visited.count(i) == 0)
        {
            depth_first_search(adjacency_list, i, visited);
            components_count++;
        }
    }
    
    return components_count;
}
```

Time complexity is O(n+m) as we need to traverse all the nodes and edges. Space complexity is O(n+m) for the adjacency list, dfs is O(n) in the worst case(graph height) and visited set is also only O(n).

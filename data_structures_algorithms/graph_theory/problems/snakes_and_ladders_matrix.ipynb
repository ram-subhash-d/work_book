Find the minimum number of die rolls necessary to reach the final cell of the given Snakes and Ladders board game.
Rules are as usual. Player starts from cell one, rolls a die and advances 1-6 (a random number of) steps at a time. Should they land on a cell where a ladder starts, the player immediately climbs up that ladder. Similarly, having landed on a cell with a snake’s mouth, the player goes down to the tail of that snake before they roll the die again. Game ends when the player arrives at the last cell.

The function has two input arguments:n, size of the board, and moves, array of integers defining the snakes and ladders: moves{i} = -1: no ladder or snake starts at i-th cell, moves{i} < i: snake from i down to moves{i}, moves{i} > i: ladder from i up to moves{i}. The indices and values in moves array are zero-based, for example moves{1} = 18 means there is a ladder from cell 2 up to cell 19.  
Example :- Input : {"n": 20,"moves": {-1, 18, -1, -1, -1, -1, -1, -1, 2, -1, -1, -1, 15, -1, -1, -1, -1, -1, -1, -1}} -> Output : 2.

## Solution : Build adjacency list and apply BFS to get shortest path
We build the directional adjacency list based on the moves that are possible from a node, the we do BFS to find the shortest path between the source and the destination, to get the minimum moves to reach the final cell.

```C++
int shortest_path(int n, vector<vector<int>> &adj_list, int src, int dst)
{
    //Set to keep track of the the visited nodes
	unordered_set<int> visited;
	vector<int> dist;
	dist.resize(n, -1);
	
    //Queue for bfs and push the src
	queue<int> level_q;
	level_q.push(src);
	visited.insert(src);
	dist[src] = 0;
	
	while(!level_q.empty())
	{
		int node = level_q.front();
		level_q.pop();
		if(node == dst)
		{
			return dist[node];
		}
		
		for(int neigh : adj_list[node])
		{
		    if(visited.count(neigh) == 0)
			{
				level_q.push(neigh);
				visited.insert(neigh);
				dist[neigh] = dist[node] + 1;
			}
		}
	}
	
	return -1;
}


int minimum_number_of_rolls(int n, vector<int> &moves) 
{
    // Write your code here.

    //Build the adjacency list
	vector<vector<int>> adj_list(n);
	for(int i = 0; i < n; i++)
	{
		for(int j = i + 1; j <= i + 6; j++)
		{
		    if(j < n)
			{
			    if(moves[j] == -1)
				{
					adj_list[i].push_back(j);
				}
				else
				{
					adj_list[i].push_back(abs(moves[j]));
				}
			}
		}
	}
	
	//Do BFS and get the shortest path
	int steps = shortest_path(n, adj_list, 0, n-1);
    return steps;
}
```

Time complexity is O(m * n) as we are building the adjacency list, space compexity will also be O(m * n) for the adjacency list.

Given a two-dimensional grid of 0s and 1s, find the size of the largest island. If there is no island return 0.An island is a group of 1s connected vertically or horizontally.  
Example :- Input : {"grid": {{1, 1, 0},{1, 1, 0},{0, 0, 1}}} -> Output : 4.

## Solution : Recursion(DFS)
We can recursively traverse through cells in the matrix(DFS) to identify islands and thier sizes, and keep track of the max size, there is no need to convert the matrix into adjacency list.

```C++
void size_island(vector<vector<int>> &matrix, vector<vector<int>> &visited, int row, int col, int rows, int cols, int &size)
{
	visited[row][col] = 1;
	
	if(col+1 < cols && visited[row][col+1] != 1 && matrix[row][col+1] == 1)
	{
	    size++;
		size_island(matrix, visited, row, col + 1, rows, cols, size);
	}
	if(col-1 >= 0 && visited[row][col-1] != 1 && matrix[row][col-1] == 1)
	{
	    size++;
		size_island(matrix, visited, row, col - 1, rows, cols, size);
	}
	if(row+1 < rows && visited[row+1][col] != 1 && matrix[row+1][col] == 1)
	{
		size++;
		size_island(matrix, visited, row + 1, col, rows, cols, size);
	}
	if(row-1 >= 0 && visited[row-1][col] != 1 && matrix[row-1][col] == 1)
	{
		size++;
		size_island(matrix, visited, row - 1, col, rows, cols, size);
	}
}

int max_island_size(vector<vector<int>> &grid) 
{    
    int rows = grid.size();
	int cols = grid[0].size();
	int max_count = 0;
	
	//Array to keep track of the visited cells
	vector<vector<int>> visited;
	visited.resize(rows);
	for(int i = 0; i < rows; i++)
	{
	    vector<int> visited_col;
	    visited_col.resize(cols, -1);
	    visited[i] = visited_col;
	}
	
	//Loop through the matrix and get the max size of island
	for(int row = 0; row < rows; row++)
	{
		for(int col = 0; col < cols; col++)
		{
			if(grid[row][col] == 1 && visited[row][col] != 1)
			{
				int size = 1;
				size_island(grid, visited, row, col, rows, cols, size);
				max_count = max(size, max_count);
			}
		}
	}
	
    return max_count;
}
```

Time complexity is O(m * n) as we are going to traverse the entire matrix, space compexity will also be O(m * n) as in the worst case when the entire matrix is 1's the call stack size will be m * n.

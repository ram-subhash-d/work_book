There are n people living in a town. Some of them dislike each other. Given the value of n and two equal length integer arrays called dislike1 and dislike2. For each i in {0, dislike1_size - 1}, the person dislike1{i} dislikes the person dislike2{i}. Check if we can divide the people of the town into two sets such that each person belongs to exactly one set and no two persons disliking each other belong to the same set.
Example :- Input : {"num_of_people": 5,"dislike1": {0, 1, 1, 2, 3},"dislike2": {2, 2, 4, 3, 4}} -> Output : 1.

## Solution : DFS/BFS
To solve this we can use a color array, marking and parent and neighbours into seperate groups, during traversal anytime if parent and neighbour fall in the same group, then we cannot seperate them. We can use either BFS or DFS.

```C++
bool is_odd_cycle(vector<vector<int>> &adj_list, queue<int> &level_q, vector<int> &color, unordered_set<int> &visited)
{
	while(!level_q.empty())
	{
		int node = level_q.front();
	    level_q.pop();
		
		for(int neigh : adj_list[node])
		{
			if(visited.count(neigh) == 0)
			{
				level_q.push(neigh);
				visited.insert(neigh);
				color[neigh] = !color[node];
			}
			else if(color[neigh] == color[node])
			{
				return true;
			}
		}
	}
	
	return false;
}

bool can_be_divided(int num_of_people, vector<int> &dislike1, vector<int> &dislike2) 
{   
    vector<vector<int>> adj_list(num_of_people);
	for(int i = 0; i < dislike1.size(); i++)
	{
		adj_list[dislike1[i]].push_back(dislike2[i]);
		adj_list[dislike2[i]].push_back(dislike1[i]);
	}
	
	unordered_set<int> visited;
	vector<int> color{};
	color.resize(num_of_people);
	for(int i = 0; i < color.size(); i++)
	{
	    color[i] = -1;
	}
	
	//Queue to do a BFS and insert the first element
	queue<int> level_q;
	bool res = false;
	
	for(int i = 0; i < num_of_people; i++)
	{
		if(visited.count(i) == 0)
		{
			level_q.push(i);
			visited.insert(i);
			color[i] = 0;
			res = !is_odd_cycle(adj_list, level_q, color, visited); 
			if(!res)
			{
				break;
			}
		}
	}
	
    return res;
}
```

Time complexity is O(n+m) as we need to traverse all the nodes and edges. Space complexity is O(n+m) for the adjacency list.

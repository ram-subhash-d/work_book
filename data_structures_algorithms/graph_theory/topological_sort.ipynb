The concept of finding the order of traversal in a directed graph is known as topological sorting.
To find a cycle in a directed graph we have 2 approaches. 
* Using DFS : It is known as arrival/departure concept.
* Using BFS : It is known as kanns algorithm.  

Arrival/Departure : We keep track of the arrival and departure for each node as we traverse the tree in DFS. If a node has its arrival marked and departure not marked and if we arrive again then the graph has a cycle in it, this new arrival edge is called the back edge.   
In get the order of the traversal push the elements into a vector as they are processed in DFS and reverse the vector to get the order, as DFS processes leaf elements first.  

Kann's Algorithm : In BFS we use indegree of each node to traverse and identify cycles. Nodes which have 0 indegree are independent nodes. We start with the independent nodes, push them into the queue, as we remove them from the queue, we remove the edges from those nodes, this may result in new independent nodes, these should be added to the queue. This way we should be able to push and pop all the nodes of the graph(then no cycle in the graph), the topological sorting order is the pop order.

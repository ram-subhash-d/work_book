## General traversal
If a graph is not fully connected then the graph will have multiple components in it. If we have to find if a graph is fully connected, we have to start at some vertix and traverse from there and check if we are able to reach all vertices. If the traversed vertices are not equal to the total vertices then the graph has components. In general to traverse a tree 
* We use a captured vertices set, to keep track of the vertices that are already processed.
* We use a visited vertices set, to keep track of the vertices that are already visited, this way we will not loop through the same vertices
* And a parents array, to keep track of the parent of each vertix in our traversal, this is called the search tree, from it we can get the path between any 2 nodes of the connected component.

All the graph algorithms will follow this basic template, where they are going to differ is the policy that they use to decide which neibour to traverse next, this will result in different search trees and properties. Below are some of the well known algorithms, their policy and the search tree they produce.
* Breath first search : It will traverse through all the neibour's of the current node before it proceeds to one of the neibours to continue the same process, it will result in BFS search tree.
* Depth first search : It will traverse through all the first neibours in a path before it proceeds to the next path, it will result in DFS search tree.
* Dijkstra's : Here we are going to traverse to the neibour which has the smallest numerical label, the vertices will have numerical labels. Here we get the shortest path search tree.
* Prim's : Same as Dijkstra's, they only differ in the numerical label's and what they mean. It will yield a minumum spanning tree.
* Best first search : Same as Dijkstra's, they only differ in the numerical label's and what they mean. It will yield a best first search tree.
* A<sup>*</sup> : Same as Dijkstra's, but there are 2 label's here they need to be added and traverse to the neibour with the smallest sum. We get a A<sup>*</sup> tree.

## Breadth first search(BFS)
Explore the graph in increasing order of distance from S. First capture the immediate neibours of S(one hop away), then capture their neighbors(2 hops away) and so on. In this traversal the path from the source to any given node will be the shortest.  
To do a BFS we use a queue, we insert the source neibhour's into the queue, as we process these we insert neighbors 2 hops away to be processed next into the queue and so on. As with all graph algorithms we will also need a captured vertices set, visited vertices set and parents array.  
The edges used in the resultant BFS search tree are called tree edges and the onces not used are called cross edges. If a graph has cross edges that means it has cycles in it. Cross edges can be at the same level or between 2 consecutive levels.  
The time complexity is O(m + n).

```C++
void bfs(vector<vector<int>> &adj_list, vector<int> &captured, unordered_set<int> &visited, queue<int> &level_q)
{
    //Do a level order traversal of a vertix
    while(!level_q.empty())
    {
        //Process the item
        int node = level_q.front();
        level_q.pop();
        captured.push_back(node);
        
        //Add the node neighbours to the queue
        for(int neigh : adj_list[node])
        {
            if(visited.count(neigh) == 1)
            {
                continue;
            }
            
            level_q.push(neigh);
            visited.insert(neigh);
        }
    }
}

vector<int> bfs_traversal(int n, vector<vector<int>> &edges) {
    // Write your code here.
    
    //Create the adjacency list
    vector<vector<int>> adj_list(n);
    for(vector<int> &edge : edges)
    {
        adj_list[edge[0]].push_back(edge[1]);
        adj_list[edge[1]].push_back(edge[0]);
    }
    
    //Variables to store the result
    vector<int> captured;
    unordered_set<int> visited;
    
    //Queue to perform a level order traversal and push the first element
    queue<int> level_q;
    
    //Process node by node to do the traversal
    //Components will be traversed
    for(int i = 0; i < n; i++)
    {
        //If node is visited continue
        if(visited.count(i) == 1)
        {
            continue;
        }
        
        //Push into queue and traverse
        level_q.push(i);
        visited.insert(i);
        bfs(adj_list, captured, visited, level_q);
    }
    
    return captured;
}
```

## Depth first search(DFS)
Instead of going broad go as deep as possible and only retreat/backtrack when neccessary. If G represented a maze, think of how you would try to get out.  
To do a DFS we use a stack to process the most recent element, here we insert the source neighbours into the stack, first neighbour's neighboirs into the stack and so on. This way we process elements depthwise first.  
We can the take the BFS algorithm and replace the queue with stack, then it will work as DFS. As with BFS time complexity is O(m + n).  
But normally DFS is implemented using recursion. In recursion, visited and captured sets are the same, this is because in DFS we capture the node as soon as we see it, so we will need a visited vertices set and parents array.

```C++
void dfs_helper(vector<vector<int>> &adj_list, int node, vector<int> &visited)
{
    visited.push_back(node);
    
    //Recursively call the neighbours
    for(int neigh : adj_list[node])
    {
        if(find(visited.begin(), visited.end(), neigh) == visited.end())
        {
            dfs_helper(adj_list, neigh, visited);
        }
    }
}

vector<int> dfs_traversal(int n, vector<vector<int>> &edges) 
{   
    //Construct the adj_list
    vector<vector<int>> adj_list(n);
    for(vector<int> &edge : edges)
    {
        adj_list[edge[0]].push_back(edge[1]);
        adj_list[edge[1]].push_back(edge[0]);
    }
    
    //Visited vector to store the result
    vector<int> visited;
    
    //Loop through the nodes and traverse the trees
    for(int i = 0; i < n; i++)
    {
        if(find(visited.begin(), visited.end(), i) == visited.end())
        {
            dfs_helper(adj_list, i, visited);
        }
    }
    
    return visited;
}
```

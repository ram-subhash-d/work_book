Graph theory was discovered by Leonhard Euler, when he wanted to solve Konigsberg bridge problem. Konigsberg had 7 bridges, he wanted to know if a path existed that crossed each bridge only once. To know if a problem can be solved with graphs, we have to check if the problem can be modeled as a graph.  

A graph has vertices also called nodes. The vertices are connected by lines known as edges. If the edges don't have any direction on them, it is called a undirected graph, then each edge is the same in either of the directions, i.e edge CD is same as edge DC. If the edges have direction on them, then it is called a directed graph, movement on the edge can only happen in the given direction. Two vertices are said to be adjacent if there is a edge directly linking them. Path in a graph is a sequence of adjacent vertices. Cycle is path which ends at the same vertix from where it started. Degree of a vertix is the number of vertices adjacent to the given vertix. If we have a graph where there are multiple edges between same two vertices, then it is called a multi graph.  

Graph is represented by G(V, E), where V is the set of all vertices and E is the set of all edges. The sum of the degrees of all the vertices in an undirected graph G = (V, E) is 2E, an important property we use while solving graph problems. In case of directed graphs there are 2 notions of degrees, out-degree and in-degree of a vertix. Therefore, the sum of the out-degrees of all the vertices in an directed graph G = (V, E) is E, it is the same for sum of the in-degrees of all the vertices.



## Graph representation
The variable n denotes the number of vertices in the graph, |V| and varaible m denotes the number of edges in the graph, |E|.
There are 4 ways to represent graphs on a computer.
* Edge lists : It can be represented as a list of edges. Each edge will have information about the edge (like weight, representing distance between the 2 verices) and vertix1 and vertix2. There might be a seperate list for vertices, edge will refer to the items in the vertices list either directly or by ID. These lists need not be ordered. The complexity of a graph is determined by the number of edges, a simple graph with n vertices can have a max of nC<sub>2</sub> edges, which is n(n-1)/2, therefore O(n<sup>2</sup>) edges. Therefore all graph representations primarily use edges for representation. None the less the vetices list can be stored in a hashmap to easily get a vertix given its ID(in O(1) time). In this kind of representation traversing the graph is time comsuming as we need to go through all the edges to identify the edges coming out of a given vertix. For space complexity we need O(n + m).
* Adjacency list : Here we store each edge with its end point vertices, i.e each vertix will have a list of vertices it connects to. This way we can easily traverse the neibouring nodes in O(1) time. If the vertix id is a convinient interger then the base vertix structure can a array, if the ids are strings we have to use hashmaps. Space complexity will still be O(m + n).
* Adjacency matrix : Here we have a n x n matrix(n is number of vertices). If 2 vertices have a edge then there corresponding cell will have a 1, otherwise a 0. For example if vertix 0 and 6 are connected then row 0 and column 6 will be a 1, if undirected row 6 and column 0 will also be a 1. For any vertix if we have to find all the neibours time complexity will be O(n), as we need to traverse through all the columns for that vertix's row in the matrix. Space complexity will be O(n<sup>2</sup>). If the graph is dense (has lots and lots of edges, |E| ~ |V|<sup>2</sup>), then this representation is good for space as it is always n<sup>2</sup>. If it is a sprase graph then adjacency list is the best fit. Since most real world graphs are sparse, we mostly use adjacency list.
* Adjacency Maps : Its a variation of adjacency list. Instead of having a list of vertices the vertix connects to, it has a map of vertices to edges the vertix connects to. This allows you to not only iterate over the neighbors in constant time, but also find if a vertix is a neibour of another given vertix in constant time, since the adjacent vertices are also stored in a hashmap. This lookup to check if 2 vertices are neibours is also a constant time in  adjacency matrix, so in some sense adjacency map combines the advantages of adjacency lists and adjacency matrices.

## Convert edge list To adjacency list
```C++
vector<vector<int>> convert_edge_list_to_adjacency_list(int n, vector<vector<int>> &edges) 
{
    
    vector<vector<int>> adjacency_list;
    
    //make n adjacency lists
    for(int i = 0; i < n; i++)
    {
        //build and push the list for vertix i
        vector<int> v_list;
        for(vector<int> &edge : edges)
        {
           if(edge[0] == i)
           {
               v_list.push_back(edge[1]);
           }
           else if(edge[1] == i)
           {
               v_list.push_back(edge[0]);
           }
        }
        std::sort(v_list.begin(), v_list.end());
        adjacency_list.push_back(v_list);
    }
    
    return adjacency_list;
}
```

## Convert edge list to adjacency matrix
```C++
vector<vector<bool>> convert_edge_list_to_adjacency_matrix(int n, vector<vector<int>> &edges) 
{   
    vector<vector<bool>> adjacency_matrix;
    for(int i; i < n; i++)
    {
        adjacency_matrix.push_back(vector<bool>(n, false));
    }
    
    //Fill up columns and rows which have edges
    for(vector<int> &edge : edges)
    {
        adjacency_matrix[edge[0]][edge[1]] = true;
        adjacency_matrix[edge[1]][edge[0]] = true;
    }
    
    return adjacency_matrix;
}
```

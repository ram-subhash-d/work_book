Given a m X n 2D grid, count the number of unique path from source(0, 0) to destination(m-1,n-1), right and down are the only 2 moves allowed. Every time we make a move either the row number will increase by 1 or the column number will increase by 1. Total number of moves will be m + n - 2, m - 1 down movements and n - 1 right movements for any valid path.  
This is same as number of ways to pick n-1 students from the total of m + n -2 students(counting subsets). If we pick one n - 1 combination, then this n - 1 moves plus the remaining m - 1 moves will give 1 path, thus the total number of ways to form n - 1 subsets from m + n - 2 elements will give the number of unique paths from source to destination. This is same as <sup>m+n-2</sup>C<sub>n-1</sub>(m+n-2 choose n-1) or <sup>m+n-2</sup>C<sub>m-1</sub>(m+n-2 choose m-1).

To solve this problem with DP, think of it in decrease and conquer strategy like the climbing n steps problem. The number of unique paths to cell m-1, n-1 is equal to sum of unique paths to the cell above it plus sum of unique paths to the cell left of it.  
f(m-1, n-1) = f(m-2, n-1) + f(m-1, n-2)  
There will be a total of m * n subproblems. We have to solve the problem in topological sort order, to solve the problem using DP, this way when we are calculating for a cell its dependent cells are already solved, the topololical sort order for this problem is left to right and top to bottom. The base cases would be the 0th row and 0th column, they will be 1's, as they is only oneway to get to these cells, 0th row have no cells above them, 0th column have no cells right of them.

```
def count_paths(m,n):
    table = array[m][n]
    
    //Base cases
    //Fill the first column with 1's
    for i in 0 to m-1
        table[i][0] = 1
    //Fill the first row with 1's
    for i in 0 to n-1
        table[0][i] = 1
        
    //Traverse in topological sort order and solve the higher order problems
    for row in 1 to m-1:
       for col in 1 to n-1:
           table[row][col] = table[row-1][col] + table[row][col-1]
           
    return table[m-1][n-1]
```
As we are traversing the full grid once, the time complexity would be O(m * n). The space complexity is also O(m * n). We can improve the space complexity by using just 2 rows, since the values in the current row are only dependent on the values in the previous row. Below is the C++ implementation.  
```C++
int unique_paths(int n, int m) 
{
    // Write your code here.
    
    //Table for DP
    vector<vector<int>> table;
    for(int i = 0; i < n; i++)
    {
        vector<int> col;
        col.resize(m, 0);
        
        table.push_back(col);
    }
    
    //Base cases fisrt row and first columns to 1
    for(int col = 0; col < m; col++)
    {
        table[0][col] = 1;
    }
    for(int row = 0; row < n; row++)
    {
        table[row][0] = 1;
    }
    
    //Build higher order problems using DP
    for(int row = 1; row < n; row++)
    {
        for(int col = 1; col < m; col++)
        {
            table[row][col] = table[row-1][col] + table[row][col-1];
        }
    }
    
    return table[n-1][m-1];
}
```

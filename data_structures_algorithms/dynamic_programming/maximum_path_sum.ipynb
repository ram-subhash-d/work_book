This is optimization problem not a counting problem. Given a m x n grid and each cell has a value, find the path from 0,0 to m-1,n-1 with the maximum value(sum the values in the cells). Say some S --> E is the best path between source S to a node E, then this will also be the best path for any intermitant node T on this path, S --> T. This property is called the optimal sub structure property. Using this property if we know the optimal paths to the cell above and left, we can calculate the optimal path to the destination cell by adding its value. We can use DP to avoid repetitive calculations. We again need to visualize a dependency graph on the grid to solve the problem with DP, the topological sort order will again be row by row. Base cases are the vertices in the 0th row and 0th column, there is only one legitimate path to each of these base case vertices.  
```
def maxpath(grid):
{
    //DP table, will keep the max paths to that cell
    table = array[m][n]
    
    //Base cases
    table[0][0] = grid[0][0]
    //Fill the base case values for row 0
    for j in 1 to n-1:
        table[0][j] = table[0][j-1] + grid[0][j]
    //Fill the base case values for col 0
    for i in 1 to m-1:
        table[i][0] = table[i-1][0] + grid[i][0]
        
    //Now perform DP in topological sort order, row by row
    for row in 1 to m-1:
        for col in 1 to n-1:
            table[row][col] = max(table[row-1][col], table[row][col-1]) + grid[row][col]
            
    return table[m-1][n-1]
}
```
The time complexity of this algorithm will be O(m * n) as we are traversing the entire grid. The space complexity is also O(m * n). If we also want the best path along with the value, then in each node we also need to save the neighbour, whose value is max of the 2 neighbours, then we can back track the path from destination to source to get the best path.

Here is the C++ implementation
```C++
int maximum_path_sum(vector<vector<int>> &grid) 
{    
    int rows = grid.size();
    int cols = grid[0].size();
    
    //Create a table for DP, store max value
    vector<vector<int>> table;
    for(int row; row < rows; row++)
    {
        vector<int> col;
        col.resize(cols, 0);
        
        table.push_back(col);
    }
    
    //Base cases
    table[0][0] = grid[0][0];
    //Fill up 0th row
    for(int col = 1; col < cols; col++)
    {
        table[0][col] = table[0][col-1] + grid[0][col];
    }
    //Fill up the 0th col
    for(int row = 1; row < rows; row++)
    {
        table[row][0] = table[row-1][0] + grid[row][0];
    }
    
    //Fill up rest of the table from the base cases
    for(int row = 1; row < rows; row++)
    {
        for(int col = 1; col < cols; col++)
        {
            table[row][col] = max(table[row-1][col], table[row][col-1]) + grid[row][col];
        }
    }
    
    
    return table[rows-1][cols-1];
}
```

Given a rod of length 'L' and a prices array where prices{i} is equal to selling a rod of length 'i'. Determine the maximum price of selling the rod of length 'L' obtainable by cutting the rod anywhere between {0...L} as many times.  
Example :- Input : L = 6, prices = {0, 1, 3, 5, 4, 7, 10} -> Output : 10.

## Recursive Implementation
This is a knapsack problem, we have to select a subset to get maximum price. For a rod of length i, try all possible cuts j (1 ≤ j ≤ i). For each cut, add the price of length j with the best profit of remaining rod (i − j).Take the maximum profit among all possible cuts. Below is the recursive implementation.

```C++
int max_comb(int i, vector<int>& cuts)
{
    // Base case
    if (i == 0)
    {
        return 0;
    }

    int ans = 0;

    // Find maximum value for rod of length i
    // by considering each cut of length j 
    // such that j <= i
    for (int j = 1; j <= i; j++) 
    {
        ans = max(ans, cut[j] + max_comb(i - j, cut));
    }

    return ans;
}

int maxCost(int n, vector<int>& cuts) 
{
    return max_comb(n - 1, cuts);
}
```

## DP Solution
To convert this into a DP solution
* Only i, remaining rod length is changing, we can use 1D array.
* Length of this array should be L+1(0 to L)
* Direction : 0 --> L
* Copy the recursion logic

```C++
int max_price(int n, vector<int>& cuts)
{
    //Price DP array
    vector<int> dp;
    dp.resize(n+1, 0);
    
    //DP
    for(int i; i <= n; i++)
    {
        int max_price = 0;
        for(int j =0; j <= i; j++)
        {
            max_price = max(max_price, cuts[j] + dp[i-j]);
        }
        dp[i] = max_price;
    }
    
    return dp[n];
}
```
Time complexity is O(L<sup>2</sup>) and space complexity is O(L).

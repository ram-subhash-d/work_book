Given a 2D array of integers representing width and height of an envelope, find the mazimum number of envelopes that you can stack inside each other.  
Example :- Input : envelopes : {{5, 4}, {6, 4}, {6, 7}, {2,3}} -> Output : 3, (envelope {2,3} inside envelope {5,4} which again can fit inside envelope {6,7}).

We need both width and height less for an envelope to fit into another envelope and we cannot directly apply longest increasing subsequence(LIS) over 2 dimensions. We can sort the arrays by width and then apply the LIS on height to get the answer, in sorting if widths are same then heights should be in desending order. This solution can be used accross all stacking problems like box stacking, weighted scheduling varaints in real world etc.

```C++
int maxEnvelopes(vector<vector<int>>& envelopes) 
{
    //sort the envelopes in the increasing order of their widths
    auto compare_by_width = [](vector<int> &a, vector<int> &b)
                            {
                                if(a[0] == b[0])
                                {
                                    return a[1] > b[1];
                                }

                                return a[0] < b[0];
                            };
    sort(envelopes.begin(), envelopes.end(), compare_by_width);

    //Do LIS on the heights to get the max stack length
    vector<int> dp;
    dp.resize(envelopes.size(), 1);
    int max_length = 1;

    for(int i = 0; i < envelopes.size(); i++)
    {
        for(int j = 0; j < i; j++)
        {
            if(envelopes[j][1] < envelopes[i][1])
            {
                dp[i] = max(dp[i], 1 + dp[j]);
            }
            max_length = max(dp[i], max_length);
        }
    }

    return max_length;
}
```
Time Complexity is O(nlogn)(for sorting) + O(n<sup>2</sup>)(for LIS), space complexity is O(n) for the dp array.

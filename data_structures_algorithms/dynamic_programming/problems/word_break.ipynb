Given a string and a dictionary of words, return true if the string can be segmented into valid dictionary words.  
Example :- Input : S = "leetcode code leet", words = {"leet", "code"} -> Output : true(we can break the complete string into only words present in the dictionary).  
Example 2:- Input : S = "catsandog", words = {"cat", "dos", "cats", "and", "sand"} -> Output : false(no possible break to match words in the dictionary).  

## Recursive Implementation
To solve this problem at each character we have a choice of breaking or not breaking. We need to keep track of the previous word end index and current index, to find if the word is present in the dictionary.

```C++
bool is_word_break(string s, vector<string>& word_dict, int i)
{
    if(i == s.size())
    {
        return true;
    }

    for(int j = i +1; j < s.size(); j++)
    {
        string word = s.substr(i, j-i+1);
        if(find(word_dict.begin(), word_dict.end(), word) != word_dict.end())
        {
            if(is_word_break(s, word_dict, j+1))
            {
                return true;
            }
        }
    }

    return false;
}

bool wordBreak(string s, vector<string>& wordDict) 
{
    return is_word_break(s, wordDict, 0);
}
```

## Bottom-up DP solution
Can we use DP here? We are solving the same subproblems on different tree branches by making different decisions in the past, i.e we can land on the same sub string from different paths, the solution from that index onwards will be the same, hence it can be solved with DP.
* Data structure : Only one variable is changing, i, so we can use a 1D array.
* Size : 0 to size of string, so n + 1 length array needed
* Direction : Low to high

```C++
bool wordBreak(string s, vector<string>& wordDict) 
{
    //Initiakize the DP
    vector<bool> dp;
    dp.resize(s.size() + 1, false);

    //Base condition, empty string is always valid
    dp[0] = true; 

    //Move backwards and fill the DP
    for(int i = 1; i <= s.size(); i++)
    {
        for(int j = 0; j <= i - 1; j++)
        {
            string word = s.substr(j, i - j);
            if(find(wordDict.begin(), wordDict.end(), word) != wordDict.end())
            {
                if(dp[j] == true)
                {
                    dp[i] = true;
                    break;
                }
            }
        }
    }

    return dp[s.size()];
}
```

Time complexity now is O(n<sup>2</sup>) and the space complexity is also O(n).

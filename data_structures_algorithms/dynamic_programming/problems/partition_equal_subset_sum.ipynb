Given an array of integers, return true if the array can be split into 2 partitions such that the sum of the partitions are equal. Partion is a subset here it does need to be continous.  
Example :- Input : arr : {1, 5, 11, 5} -> Output : true, (partition as {1,5,5} and {11}).

## Recursive Implementation
If we can identify a subset whose sum is equal to total array sum/2, then it means we can partition the array into 2 arrays of same sum. So this becomes a subset problem whose sum is total array sum/2. Since it is a subset problem, at each recursion we have the option of including or excluding a number.

```C++
bool is_partial_sum(vector<int>& nums, int partial_sum, int cur_sum, int i)
{
    if(cur_sum == partial_sum)
    {
        return true;
    }

    if(i == nums.size())
    {
        return false;
    }

    return is_partial_sum(nums, partial_sum, cur_sum + nums[i], i+1) || is_partial_sum(nums, partial_sum, cur_sum, i+1);
}

bool canPartition(vector<int>& nums) 
{
    //Get total sum
    int total_sum = accumulate(nums.begin(), nums.end(), 0);

    //If total sum is odd we cannot partion into 2 equal sums
    if(total_sum%2 != 0)
    {
        return false;
    }

    return is_partial_sum(nums, total_sum/2, 0, 0);
}
```

## Bottom-up DP solution
* Data structure : i and cur_sum are continuously changing, so we need a 2D array for DP
* Dimension : N+1 for i and T+1 for cur_sum, T is target sum.
* Direction : 0 --> N and 0 --> T
* Copy paste

```C++
bool canPartition(vector<int>& nums) 
{
    //Get total sum
    int total_sum = accumulate(nums.begin(), nums.end(), 0);

    //If total sum is odd we cannot partion into 2 equal sums
    if(total_sum%2 != 0)
    {
        return false;
    }

    int len = nums.size();
    int target_sum = total_sum/2;
    vector<vector<bool>> dp;
    for(int i = 0; i <= len; i++)
    {
        vector<bool> col;
        col.resize(target_sum+1, false);

        dp.push_back(col);
    }

    //Base condition
    dp[0][0] = true;   //Can we create a target of 0 from 0 onwards

    for(int i = 1; i <= len; i++)
    {
        int cur_sum = nums[i-1];
        for(int j = 0; j <= target_sum; j++)
        {
            if(j < cur_sum)   //avoid picking if j < cur_sum
            {
                dp[i][j] = dp[i-1][j];
            }
            else //2 options pick or not pick
            {
                dp[i][j] = dp[i-1][j-cur_sum] || dp[i-1][j];
            }
        }
    }

    return dp[len][target_sum];
}
```

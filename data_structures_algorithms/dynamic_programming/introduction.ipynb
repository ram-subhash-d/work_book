Dynamic programming is recursion without repetition, it is a smart way of doing recursion without doing repetitive or redundant work. For example if we take the fibonacci recursive function. So a recursive problem with overlapping subproblems can be solved with DP. 

```
def fib(n):
    if n == 0 or n == 1
        return n
    return fib(n-1) + fib(n-2)
```
The execution of this recursive function will be a recursive tree, in this recursive tree the same sub problems appear many times. For example fib(n - 4) will appear many many times, so there will be a lot of redundant work going on. This naive recursive implementation took exponential time because of the repeated work. Avoiding the repeated work should make the execution much faster.

## Top Down Memoization
The recursive function above results in DFS, solving the problem in a top down manner(moving from bigger to smaller problems). In the recursive tree we will be calculating the value for the same n multiple times, recursively. What if we cache/store the value, the first time we calculate for any particular n and use it for the other times to remove repetitive work? We can store it in a hash map, for example for fib(2), key will be 2 and value will value of fib(2). So the next time the recursive tree comes across fib(2), we can use the value in the hashmap and return, instead of recursively recalculating it. This way of storing the previous results to reuse is known as memoization.

```
def fib(n):
    if n in memo:
        return memo[n]
    if n == 0 or n == 1
        return n
    memo[n] = fib(n-1) + fib(n-2)
    return memo[n]
```
If we store the base case values in the memo before calling the recursive function, we can remove the base case condition in the recursive function.  
This will result in O(n) nodes in the recursive tree instead of exponential number of nodes. So the time complexity is O(n), for the memo we need an extra O(n) space, so we gained time by using extra space.

## Bottom Up
This is the more preferred approach for dynamic programming, as it avoids function's call stack memory. For this approach we to convert recursive tree into a dependency graph, by having edges from sub problems to problems and colapsing nodes that repeat.
![rt_to_dg](rt_to_dg.png)  

This dependency graph shows the dependency between different subproblems, for example f(4) depends on the computation of f(3) and f(2). This dependency graph is directed and acyclic, f(1) and f(0) are independent nodes which form the base cases. Hence we can solve the problem in topological sort order, starting from base cases and building for the bigger problems. If we caculate the values of f in topological sort order and store we can avoid repetitions, for example if we come to f(3) in topological sort order we would have already caculated the sub problems that it depends on, f(2) and f(1).  
```
def fib(n):
    table = new of size n+1
    table[0] = 0
    table[1] = 1
    for i in 2 to n:
        table[i] = table[i-1] + table[i-2]
        
    return table[n]
```  
Both time and space complexity are O(n).


## Fibonacci Number
We can also improve the space complexity for this problem by just using 3 elements in the array.  

```C++
int find_fibonacci(int n) 
{    
    if(n == 0 || n == 1)
    {
        return n;
    }
    
    array<int, 3> tab;
    tab[0] = 0;
    tab[1] = 1;
    for(int i = 2; i <= n; i++)
    {
        tab[i%3] = tab[(i-1)%3] + tab[(i-2)%3];
    }
    
    return tab[n%3];
}
```

## Climbing n stairs
In oder to apply DP, we must be able to go from the counting problem to recurrence equation, then we have to think of how to solve the recurrence equation without doing repititions.  
The maximum number of steps he can climb at a time is 2, that is he can climb eiher 1 step or 2 steps at a time. If there are n steps in total, in how many different ways can he climb the staircase?  
If we look at this as a decrease and conquer problem and solve the problem from right, number of ways to get to step n, f(n) should be equal to, number of ways to get to step n - 1 (f(n-1)) + number of ways to get to step n - 2 (f(n-2)). So this is like the fibonacci sequence with the base conditions being f(1) = 1 and f(2) = 2.
```
def f(n):
    if n == 1 || n == 2
        return n
        
    table = new of size n+1
    table[1] = 1
    table[2] = 2
    for i in 3 to n:
        table[i] = table[i-1] + table[i-2]
        
    return table[n]
```

## Counting Subsets of Size k
We can either include a student or exclude a student from the subset at each recursion, so the recursive equation will be
c(n,k) = c(n-1, k) + c(n-1,k-1). Below is the recursive implementation.  
```
def c(n,k):
    //Base case
    if k == 0 or k == n:
        return 1
    
    //Recursive case
    return c(n-1, k) + c(n-1, k-1)
```
This recursive implementation will have a lot of repetitive work, like calculating 2 choose 1 repetitively along multiple paths. Time complexity of this recursive implementation is O(2<sup>n</sup>), exponential.  
We can use DP to avoid solving the same sub problems multiple times. There are 2 parameters here that are changing during the recursion n(0 to n) and k(0 to k). So we will have (n+1) * (k+1) subproblems, which is a small polynomial, so if can calculate a subproblem only once then the time complexity will reduce to a small polynomial, O(n * k).  
The dependency graph to solve the problem in topological sort order is as below.  
![rt_to_dg_subsets](rt_to_dg_subsets.png)  

The red nodes can be ignored, they are not part of the solution.  
```
def c(n, k):
    if k == 0 or k == n
        return 1
    
    //DP table
    table = array[n+1][k+1] //n -> 0 to n, k -> 0 to k
    
    //Base cases
    //Column 0 values are 1, k = 0
    for row in 0 to n:
        table[row][0] = 1
    //Diagnal values are 1
    for col in 0 to k:
        table[col][col] = 1
        
    //Build up the table from smaller sub problems(DP)
    for row in 2 to n: //First 2 rows are already filled in the base cases
        for col in 1 to min(row, k): //0th col filled in base case, so start at 1, till row, cap row to k 
            table[row][col] = table[row-1][col] + table[row-1][col-1]
            
    //Answer should be at the end of the table
    return table[n][k]
```
The space complexity is also O(n * k). We can improve the space complexity, looking at the dependency graph, a row depends only on its previous row, so 2 rows, space complexity would come down to O(k). Below is the C++ implementation.  
```C++
int ncr(int n, int r) 
{    
    if(r > n)
    {
        return 0;
    }
    
    //Return 1 if r = 0 or n =r
    if(r == 0 || n == r)
    {
        return 1;
    }
    
    //Initialize a 2d array of size n+1 x r+1
    vector<vector<int>> table;
    for(int i = 0; i <= n; i++)
    {
        vector<int> col;
        col.resize(r+1, 0);
        
        table.push_back(col);
    }
    
    //Base cases
    //All elements in row 0 are 1
    for(int row = 0; row <= n; row++)
    {
        table[row][0] = 1;
    }
    //All diagnal elements are 1, n = r
    for(int col = 0; col <= r; col++)
    {
        table[col][col] = 1;
    }
    
    
    //Build the upper values using the lower values(with DP)
    for(int row = 2; row <= n; row++)
    {
        for(int col = 1; col <= min(row, r); col++)
        {
            table[row][col] = table[row-1][col] + table[row-1][col-1]; 
        }
    }
    
    return table[n][r];
}
```

This is a recursive way of traversing a tree. Here we traverse the entire subtree recursively before we move to other subtrees. There are 3 kinds of DFS traversals depending on when we want to process the node, preorder, inorder and postorder. All the 3 traversals are going to take O(n) time. Space complexity will be O(h), which is O(logn) for a balanced tree and O(n) for a skewed tree.

## Preorder
Process the current node before we process any nodes in the left or the right subtrees. Useful when we want to send the data from parent to child(top-down), example like calculating the depth of a tree. For the below tree  
![sample_binary_tree](sample_binary_tree.png)  
Preorder traversal is 1,2,4,5,7,8,3,6,9.

```C++
void preorder_helper(BinaryTreeNode *node, vector<int> &res)
{
    if(!node)
    {
        return;
    }
    
    res.push_back(node->value);
    
    preorder_helper(node->left, res);
    preorder_helper(node->right, res);
}
```

## Inorder
Process all the nodes in the left subtree before we process the current node, then process the nodes in the right subtree. Useful when you want to maintain a sorted order. Exclusively used in binary search trees. For the same above tree inorder traversal will be 4,2,7,5,8,1,3,9,6.

```C++
void inorder_helper(BinaryTreeNode *node, vector<int> &res)
{
    if(!node)
    {
        return;
    }
    
    inorder_helper(node->left, res);
    
    res.push_back(node->value);
    
    inorder_helper(node->right, res);
}
```

## Postorder
Process the nodes in the left and right subtrees before we process the current node. Useful when we want to send the data from child to parent(bottom-up), example like caculating the height of the tree, calculating the space used up by directories given the size of files in a file system tree heirarchy. For the same above tree postorder traversal will be 4,7,8,5,2,9,6,3,1.

```C++
void postorder_helper(BinaryTreeNode *node, vector<int> &res)
{
    if(!node)
    {
        return;
    }
    
    postorder_helper(node->left, res);
    postorder_helper(node->right, res);
    
    res.push_back(node->value);
}
```

The same can be extended to a n-array tree, but only pre and post order traversals are possible, as a given node can have many children and we don't know after which child to process the current node for indorder traversal.

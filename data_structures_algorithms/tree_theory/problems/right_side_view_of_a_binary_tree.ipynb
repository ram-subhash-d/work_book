Given a binary tree, imagine yourself standing on the right side of it and return a list of the node values that you can see from the top to the bottom.  
Example :- Input : {"root": {0},{1, 2},{3, 4}} -> Output : {0, 2, 4}.

## Solution
Use a queue to do a BFS. Capture only the last element at each level into the result.

```C++
vector<int> right_view(BinaryTreeNode *root) 
{   
    //To the store the result
	vector<int> res;
	
	//Need a queue as we need a BFS, since we all right most elements at each level
	//And push the root node
	queue<BinaryTreeNode *> level_q;
	level_q.push(root);
	
	//Loop through the queue level by level
	while(!level_q.empty())
	{
		//Get the number of nodes at the current level
		int size = level_q.size();
		
		//Loop through all the nodes at the current level
		for(int i = 0; i < size; i++)
		{
		    BinaryTreeNode *node = level_q.front();
			level_q.pop();
			
			//Push the last element at the level into the queue
			if(i == (size - 1))
			{
				res.push_back(node->value);
			}
			
			//Push the next level nodes into the queue
			if(node->left != nullptr)
			{
				level_q.push(node->left);
			}
			if(node->right != nullptr)
			{
				level_q.push(node->right);
			}
		}
	}
	
    return {res};
}
```

The time complexity of this algorithm is O(n) as we are using a while and for loop, and traversing all the nodes. Space complexity is equal to the max number of nodes in a level(pushing them to queue), for a complete binary tree, max number of nodes is n/2 at the leaf level, hence space complexity is O(n).

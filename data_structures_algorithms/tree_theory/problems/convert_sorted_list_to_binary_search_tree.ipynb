Given a linked list with elements sorted in ascending order, convert it into a height-balanced binary search tree.  
Example :- Input : {"head": {-1, 2, 3, 5, 6, 7, 10}} -> Output : {"root": {5},{2, 7},{-1, 3, 6, 10}}.

## Solution
First we convert list to array as we can access elements in a array in O(1) time. Then we do a binary traverse of the array to process the mid element in each recursion to build a complete binary search tree.

```C++
vector<int> convert_list_to_array(LinkedListNode *head)
{
    vector<int> sorted_data;
    while(head != nullptr)
    {
        sorted_data.push_back(head->value);
        head = head->next;
    }
    
    return sorted_data;
}

BinaryTreeNode *sorted_array_to_bst(vector<int> &sorted_data, int s_i, int e_i)
{
    if(s_i > e_i)
    {
        return nullptr;
    }
    
    if(s_i == e_i)
    {
        return new BinaryTreeNode(sorted_data[s_i]);
    }
    
    int m_i = (s_i + e_i)/2;
    BinaryTreeNode *root = new BinaryTreeNode(sorted_data[m_i]);
    root->left = sorted_array_to_bst(sorted_data, s_i, m_i -1);
    root->right = sorted_array_to_bst(sorted_data, m_i + 1, e_i);
    
    return root;
}
```

The time complexity of this algorithm is O(n) as we will traverse through all the nodes in the tree, and also copy to array is O(n) time. Space complexity is O(n) as we are using a array of size n.

Given a binary tree and an integer k, check whether the tree has a root to leaf path with a sum of values equal to k.  
Example :- Input : {"root": {0},{1, 2},{3, 4},"k": 4} -> Output : true.

## Solution
Since it is a top down problem, we have to calculate the sum as we are moving through each branch from root to leaf. So we use a DFS preorder traversal.

```C++
void sum(BinaryTreeNode *node, long running_sum, int k, bool &res)
{
    if((node == nullptr) || res)
    {
        return;
    }
    
    //Process
    running_sum += node->value;
    
    //if leaf node check the sum
    if((node->left == nullptr) && (node->right == nullptr))
    {
        if(running_sum == k)
        {
            res = true;
        }
    }
    
    sum(node->left, running_sum, k, res);
    sum(node->right, running_sum, k, res);
}
```

The time complexity of this algorithm is O(n) as in the worst case we will traverse through all the nodes in the tree. Space complexity is O(h) as it is DFS, for complete binary tree it is O(logn).

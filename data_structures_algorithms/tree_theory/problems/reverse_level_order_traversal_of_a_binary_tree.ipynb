Given a binary tree, return the bottom-up level order traversal of the node values listing each level from left to right.  
Example :- Input : {"root": {0},{1, 2},{3, 4}} -> Output : {{3, 4},{1, 2},{0}}.

## Solution
Use a queue to do a BFS. Once we have the level ordered lists we have to reverse the lists to get the reverse level order. For this we can either use a temporary stack or insert the new layers at the start of the result array.

```C++
vector<vector<int>> reverse_level_order_traversal(BinaryTreeNode *root) 
{   
    vector<vector<int>> res;
	
	//Queue as we need to do a BFS and push the root element
	queue<BinaryTreeNode *> level_q;
	level_q.push(root);
	
	//Loop through the queue level by level
	while(!level_q.empty())
	{
		//Get the number of nodes at the current level
		int size = level_q.size();
		vector<int> level;
		
		//Loop through the items at level
		for(int i = 0; i < size; i++)
		{
		    BinaryTreeNode *node = level_q.front();
			level_q.pop();
			
			//Process the node at the level
			level.push_back(node->value);
			
			//Push the next level nodes into the queue
			if(node->left != nullptr)
			{
				level_q.push(node->left);
			}
			if(node->right != nullptr)
			{
				level_q.push(node->right);
			}
		}
		
		//Push the level values into the result at the start
		res.insert(res.begin(), level);
	}
	
    return {res};
}
```

The time complexity of this algorithm is O(n) as we are using a while and for and traversing all the nodes. Space complexity is equal to the max number of nodes in a level(pushing them to queue), for a complete binary tree, max number of nodes n/2 at the leaf level, hence space complexity is O(n).

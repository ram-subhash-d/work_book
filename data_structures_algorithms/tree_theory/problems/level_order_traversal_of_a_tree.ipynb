Given a tree, list node values level by level from left to right.
Example :- Input : {"root": {1, null},{3, 4, 2, null},{null, 5, 6}} -> Output : {{1},{3, 4, 2},{5, 6}}.

## Solution
Use a queue to do a BFS.

```C++
vector<vector<int>> level_order(TreeNode *root) {
    // Write your code here.
    
    //Result to hold the levels
	vector<vector<int>> res;
	
	//Queue as we need to do a BFS and push the root node into it
	queue<TreeNode *> level_q;
	level_q.push(root);
	
	//Process the queue level by level
	while(!level_q.empty())
	{
		//Get the number of nodes in the current level
		int size = level_q.size();
		vector<int> level;
		
		//Loop through the nodes at this level
		for(int i = 0; i < size; i++)
		{
			//Get the next element in the level
			TreeNode *node = level_q.front();
			level_q.pop();
			
			level.push_back(node->value);
			
			//Push the current nodes children(next level) into the queue
			for(TreeNode *ch_node : node->children)
			{
				level_q.push(ch_node);
			}
		}
		
		//Append the level into the result
		res.push_back(level);
	}
	
	
    return {res};
}
```

The time complexity of this algorithm is O(n) as we are using a while and for all and traversing all the nodes. Space complexity is equal to the max number of nodes in a level(pushing them to queue), for a complete binary tree, max number of nodes n/2 at the leaf level, hence space complexity is O(n).

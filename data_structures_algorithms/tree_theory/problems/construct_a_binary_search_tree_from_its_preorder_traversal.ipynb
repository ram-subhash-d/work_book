Construct a Binary Search Tree whose preorder traversal matches the given list.  
Example :- Input : {"preorder": {1, 0, 2}} -> Output : {"root": {1},{0, 2}}.

## Solution
In a preorder traversal the node is first added and then the left and right sub trees are processed. So given a node its left sub tree values will less than it and its right subtree values will be greater than it. We use this property and pass in the bounds for each sub tree to build the binary search tree.

```C++
int array_i = 0;

BinaryTreeNode *preorder_array_to_bst(vector<int> &preorder, int min, int max)
{
    if((array_i >= preorder.size()) || (preorder[array_i] < min) || (preorder[array_i] > max))
    {
        return nullptr;
    }
    
    BinaryTreeNode *node = new BinaryTreeNode(preorder[array_i]);
    array_i++;
    node->left = preorder_array_to_bst(preorder, min, node->value);
    node->right = preorder_array_to_bst(preorder, node->value, max);
    
    return node;
}
```

The time complexity of this algorithm is O(n) as we will traverse through all the nodes in the tree. Space complexity is O(h) as it is DFS.

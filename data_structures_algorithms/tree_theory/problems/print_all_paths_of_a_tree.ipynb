Given a binary tree, return all paths from root to leaf.  
Example :- Input : {"root": {1},{2, 3},{4, 5, 6, 7}} -> Output : {{1, 2, 4},{1, 2, 5},{1, 3, 6},{1, 3, 7}}.

## Solution
Since it is a top down problem, we use a DFS preorder traversal.

```C++
void all_paths_helper(BinaryTreeNode *node, vector<int> &slate, vector<vector<int>> &res)
{
    if(node == nullptr)
    {
        return;
    }
    
    slate.push_back(node->value);
    if((node->left == nullptr) && (node->right == nullptr))
    {
        res.push_back(slate);
    }
    
    all_paths_helper(node->left, slate, res);
    all_paths_helper(node->right, slate, res);
    
    slate.erase(slate.end() - 1);
}
```

The time complexity of this algorithm is O(n) as we will traverse through all the nodes in the tree. Space complexity is O(h) as it is DFS, for complete binary tree it is O(logn).

Given a binary tree, find its diameter.  
Example :- Input : {"root": {0},{1, 2},{3, 4}} -> Output : 3.

## Solution
Since diameter can be calculated from heights, this is a bottom up problem. So we can use DFS postorder traversal.

```C++
int get_height(BinaryTreeNode *node, int &max_diameter)
{
    if(node == nullptr)
    {
        return -1;
    }
    
    int lh = get_height(node->left, max_diameter);
    int rh = get_height(node->right, max_diameter);
    
    int diameter = lh + rh + 2;
    max_diameter = max(diameter, max_diameter);
    return max(lh, rh) + 1;
}
d
```

The time complexity of this algorithm is O(n) as we will traverse through all the nodes in the tree. Space complexity is O(h) as it is DFS, for complete binary tree it is O(logn).

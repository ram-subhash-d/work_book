## Statement
In a binary search tree, find the path between the given two nodes. Find the number of edges to traverse and/or the actual path.

## Solution
First find the edges to move from the root node to the source node, then find the edges to move from the root node to the destination node, finally remove the common edges between these two paths to get the path between the source and the destination.

```c++
struct node
{
    int value;
    node* left;
    node* right;
    
    node(int v, node* l, node* r)
    {
        value = v;
        left = l;
        right = r;
    }
}
```

```c++
#include <iostream>

class binary_search_tree
{
public:
    binary_search_tree()
    {
    }
    
    void add(int value)
    {
        node* new_node = new node(value, nullptr, nullptr);
        if(m_root == nullptr)
        {
            m_root = new_node;
            return;
        }
        add(m_root, new_node);
    }
    
    void find_path(int source, int destination)
    {
        bool res = find_path(m_root, source);
        if(res)
        {
            res = find_path(m_root, destination);
        }
        
        if(res)
        {
            std::cout << "path found::" << "steps:" << m_steps << std::endl;
            std::cout << "visited nodes:: ";
            for(int n : m_visit_nodes)
            {
                std::cout << n << " ";
            }
        }
        else
        {
            std::cout << "path not found::";
        }
    }
    
private:
    void add(node* curr_node, node* new_node)
    {
        if(new_node->value >= curr_node->value)
        {
            if(curr_node->right == nullptr)
            {
                curr_node->right = new_node;
                return;
            }
            add(curr_node->right, new_node);
        }
        else
        {
            if(curr_node->left == nullptr)
            {
                curr_node->left = new_node;
                return;
            }
            add(curr_node->left, new_node);
        }
    }
    
   bool find_path(node* curr_node, int value)
   {
       if(curr_node == nullptr)
       {
           return false;
       }
       
       if(std::find(m_visit_nodes.begin(), m_visit_nodes.end(), curr_node->value) == m_visit_nodes.end())
       {
           m_visit_nodes.push_back(curr_node->value);
           m_steps++;
       }
       else
       {
           m_steps--;
       }
       
       if(curr_node->value == value)
       {
           return true;
       }
           
       if(value > curr_node->value)
       {
           return find_path(curr_node->right, value);
       }
       else
       {
           return find_path(curr_node->left, value);
       }
   }
    
    node* m_root = nullptr;
    std::vector<int> m_visit_nodes;
    int m_steps = 0;
};

binary_search_tree bst;
bst.add(8);
bst.add(5);
bst.add(10);
bst.add(9);
bst.add(11);
bst.add(4);
bst.add(6);

bst.find_path(4, 9);
```

## Statement
Given two binary search tree's, find out if the two trees are equal.

## Solution
Do a inorder traversal of both the tree's. Inorder traversal give's us sorted output. Compare the two sorted lists to check for the equality.

```c++
struct node
{
    int value;
    node* left;
    node* right;
    
    node(int v, node* l, node* r)
    {
        value = v;
        left = l;
        right = r;
    }
}
```

```c++
#include <iostream>
#include <vector>

class binary_search_tree
{
public:
    binary_search_tree()
    {
    }
    
    void add(int value)
    {
        node* new_node = new node(value, nullptr, nullptr);
        if(m_root == nullptr)
        {
            m_root = new_node;
            return;
        }
        add(m_root, new_node);
    }
    
    void explore_inorder(std::vector<int>& inorder_nodes)
    {
        explore_inorder(m_root, inorder_nodes);
    }
    
private:
    void add(node* curr_node, node* new_node)
    {
        if(new_node->value >= curr_node->value)
        {
            if(curr_node->right == nullptr)
            {
                curr_node->right = new_node;
                return;
            }
            add(curr_node->right, new_node);
        }
        else
        {
            if(curr_node->left == nullptr)
            {
                curr_node->left = new_node;
                return;
            }
            add(curr_node->left, new_node);
        }
    }
    
    void explore_inorder(node* curr_node, std::vector<int>& inorder_nodes)
    {
        if(curr_node == nullptr)
        {
            return;
        }
        explore_inorder(curr_node->left, inorder_nodes);
        inorder_nodes.push_back(curr_node->value);
        explore_inorder(curr_node->right, inorder_nodes);
    }
    
    node* m_root = nullptr;
};

binary_search_tree bst_1;
bst_1.add(4);
bst_1.add(3);
bst_1.add(6);
std::vector<int> inorder_nodes_1;
bst_1.explore_inorder(inorder_nodes_1);

binary_search_tree bst_2;
bst_2.add(6);
bst_2.add(3);
bst_2.add(4);
std::vector<int> inorder_nodes_2;
bst_2.explore_inorder(inorder_nodes_2);

std::cout << "bst 1::";
for(int ele : inorder_nodes_1)
{
    std::cout << ele << " ";
}
std::cout << std::endl;
std::cout << "bst 2::";
for(int ele : inorder_nodes_2)
{
    std::cout << ele << " ";
}
std::cout << std::endl;

if(inorder_nodes_1 == inorder_nodes_2)
{
    std::cout << "Trees are equal";
}
else
{
    std::cout << "Trees are not equal";
}
```

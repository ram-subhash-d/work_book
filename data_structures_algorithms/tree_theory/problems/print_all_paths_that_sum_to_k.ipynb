Given a binary tree and an integer k, find all the root to leaf paths that sum to k.  
Example :- Input : {"root": {10},{25, 30},{45, null, 40, 50}, "k": 80} -> Output : {{10, 25, 45}{10, 30, 40}}.

## Solution
Since it is a top down problem, we have to calculate the sum as we are moving through each branch from root to leaf. So we use a DFS preorder traversal. As we need the paths we need to save the paths in a slate as we recurse.

```C++
void paths_sum(BinaryTreeNode *node, int k, int sum, vector<int> &slate, vector<vector<int>> &res)
{
    if(node == nullptr)
    {
        return;
    }
    
    //Pre processing
    slate.push_back(node->value);
    sum += node->value;
    if((node->left == nullptr) && (node->right == nullptr))
    {
        if(sum == k)
        {
            res.push_back(slate);
        }
    }
    
    paths_sum(node->left, k, sum, slate, res);
    paths_sum(node->right, k, sum, slate, res);
    
    slate.erase(slate.end() - 1);
}
```

The time complexity of this algorithm is O(nlogn) as we will traverse through all the nodes in the tree and the slate can have h items, copying this will take logn time for a complete binary tree. Space complexity is O(h) as it is DFS, for complete binary tree it is O(logn).

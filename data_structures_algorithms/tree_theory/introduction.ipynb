Two important classes of problems that naturally lead to tress are Dictionary and Set. Implementing them will be almost similar.
* Dictionary : A dictionary is a collection of n word meaning pairs, given a word it allows us to search for its meaning. In computer science this becomes key value pairs. We can define a dictionary abstract data type as, which stores a collection of n key value pairs and which supports a search operation by key(returns value at that key). It should be dynamic, it should support insert and delete operations.
* Set : In many cases we may want to just store keys, which is a set. A set abstract data type also supports search operation of the key.It should also be dynamic, it should support insert and delete operations.  

There main operation is search. Dictionaries are also called as maps, associative arrays and symbol tables.

## Implementing a dictionary ADT using arrays and linked lists
* Arrays : If we dump the key value pairs in a unsorted array, then search will take O(n). Insertion will be O(1), as we can append the pair at the end. Deletion needs a search and left shift of the elements after deleting the element, so O(n).  
  If we insert the key pair into the array in a sorted order, then search will be O(log n) as we can do a binary search. For insertion we need to find the location to insert which is O(log n) and also right shift the elements to make space for the element to insert which in the worst case can be O(n), so over all insertion is O(n). For deletion again we need to sreach, delete and left shift the array elements, which is O(n).
* Linked lists : If we dump the key value pairs into a linked list unsorted, since we need to check all the elements, search is going to take O(n). Insertion will be O(1), as we can append the pair at the end of the list. Deletion is also O(n) as we need to search the element and delete it.  
  If we insert the key value pair into the list in a sorted order, search will still take O(n) as we cannot find the mid element in a list in O(1), so we need to loop through the entire list in sequence. For insert we need to find the location and insert, which will again be O(n) for search. For deletion we need to again search and delete so O(n).  

If we want to implement a static dictionary with no insert and deletion, then sorted array would be the best option since search takes O(log n). But for dynamic dictionary we need a better data structure to improve the insert and delete efficiencies.

## Implementing a dictionary ADT using hash table
If we assume that keys are integers from 0 to m-1, then we can use a fixed array of size m. Then search, insert and delete are all O(1), as we need to just go to the array index(which is key) and do the respective operation. Only the amount of memory used is high as number of actual elements n can be much less than m.  
Now let us drop this assumption that keys are integers from 0 to m-1, keys can be anything(strings too), can we use this same contiguous memory solution for the general case?  
Yes, the answer is to map each key to an interger in the range 0 to m-1, in a way the keys get distributed to different indexes. This converted integer will be the array index. What we want to avoid is for 2 different keys to be mapped to the same index(avoid collisions).  
This kind of implementation where we take a fixed array of size m, and use some function(hash function) to convert keys into indexes from 0 to m-1 and store the value at that index is known as hash table. So it is combination of an array and a hash function. The hash function should be very good with minimum collisions. Some implementations allow minimum collisions by saving a small linked list at the index instead of a single value. With minimum collisions time complexity is O(1) for all 3 operations as at a given index there will be a single or a couple of values(in case of collision).  

Can we do a similar thing with linked lists instead of continuous memory array, yes with binary search tree which is described on a seperate page. Although search, insert and delete take O(1) time with hash table, operations like minimum, maximum, predecessor or successor take O(n) time as we need to seach the entire array for these values. On a balanced binary search tree all these operations from search to successor take O(logn) time.

So if a programming language is giving dictionary and set then under the hood they are implemented with hash table or a balanced binary search tree.

## Hierarchical Data
This is another context where trees are used, when your data naturally has hierarchical relationships. If we have to represent a family tree, a corporate hierarchy, unix file system, a html document or a book's table of contents, we would naturally use a tree structure to depict these hierarchical relationships. These are normally n-ary trees, where one parent can have many children. These trees do not have the ordering property like the binary search trees and they may not be balanced. The definition of this structure would be
```C++
struct tree_node
{
    int key;
    object value;
    list<tree_node*> children;
};
```

## Definitions
* Height of given node : Maximum distance from that node to any of its leaf's. In a binary tree if we know the height of the child nodes then - height(node) = max(height(node.left), height(node.right)) + 1. Height of a null node is -1.
* Depth of a node : Distance from the root of the tree. depth(chid node) = depth(node) + 1
* Binary tree : It is a tree where each node has no more than 2 nodes.
* Balanced binary tree : Every node in the tree should comform to the following property. Absolute value of height of left child minus height of right child should be less than or equal to 1. abs(lh - rh) <= 1.
* Unbalanced binary tree : If any node in the tree node does not conform to the above property, then it is an unbalanced binary tree.
* Complete binary tree : Balanced binary tree where all levels are filled from left to right. Last level may be incomplete.
* Perfect binary tree : Balanced binary tree where all levels are completely fill. For any of these balanced tree roughly we can take h = log(n), internal nodes = n/2 and leaf nodes = n/2.
* Skewed tree : Unbalanced binary tree where each node has only one child node. Here h = n and leaf nodes = 1.

A combinations/subsets is a number of ways to choose k objects out of n, where repetition is not allowed and order is also not important. Denoted by C(n, k), the number of subsets of size k that can be formed out of a set of size n. For example C(n, n) = 1, C(n, 0) = 1, C(n, 1) = n. C(n, k) = (n!)/(k! * (n-k)!), Also C(n,k) = C(n, n-k).  
One way to write C(n,k) function is to use the factorial function and caculate its value from the formula. Another way is to directly solve it by using decrease and conquer. Say for example we have to form communities of size k from a class of n students. The root node can pick one student, it can either put him in the community or not, so it is going to give C(n-1, k-1) to one child node and C(n-1, k) to the other child node to calculate and sum these 2 values when the child nodes return those values. This can be done recursively at each node to get the number of combinations. C(n, k) = C(n-1, k-1) + C(n-1, k). Base case is if k=0 or k=n return 1(C(n,0) or C(n,n) = 1). Time complexity is exponential.
```C++
int combinations(int n, int k)
{
    //Base cases
    if(k == 0 || k == n || n <= 1)
    {
        return 1;
    }

    //New smaller problems
    return combinations(n-1, k) + combinations(n-1, k-1); 
}
```

## Enumerate all Combinations/Subsets
Given a set of n distinct characters, enumerate all its subsets(all lengths possible). Every character in the input has the possibility to be included or not included in a subset, if we have the subsets for n-1 elements we can take each of those subsets and form 2 subsets, one by adding the nth element and the other by not adding the nth element, this will give the subsets for the n elements, therefore the total number of subsets possible would be 2<sup>n</sup>. Here too we can use DFS like permutations, enumerate the solution when the sub problem size becomes empty.
```C++
void sub_sets(std::string comb, std::string rem_elements)
{
    if(rem_elements.size() == 0)
    {
        std::cout << comb << std::endl;
        return;
    }

    char element = rem_elements[0];
    rem_elements.erase(0, 1);
    sub_sets(comb, rem_elements);
    sub_sets(comb + element, rem_elements);
}
```
There will be 2<sup>n</sup> leaf nodes each doing some constant amount of work and printing which can again be order n. So time complexity only for leaf nodes is θ(2<sup>n</sup>), even if we don't consider work done by other nodes this is a exponential time complexity.As it is a DFS algorithm the space complexity is θ(n).

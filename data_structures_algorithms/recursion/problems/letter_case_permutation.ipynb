Given a string, return all strings that can be generated by changing case of one or more letters in it.  
Example :- Input : {"s": "a1z"} -> Output : {"A1Z", "A1z", "a1Z", "a1z"}.

## Solution 1 : Resursion with string copy
Every character can be filled up with a upper case and a lower case. So we do upper case and lower case recusive calls for each iteration.
```C++
void permutations(string slate, string &s, int i, vector<string> &res)
{
    if(slate.size() == s.size())
    {
        res.push_back(slate);
        return;
    }
    
    if(isalpha(s[i]))
    {
        permutations(slate + static_cast<char>(toupper(s[i])), s, i+1, res);
        permutations(slate + static_cast<char>(tolower(s[i])), s, i+1, res);
    }
    else
    {
        permutations(slate + s[i], s, i+1, res);
    }
}
```

There are 2<sup>n</sup> nodes and copy is taking place at each node so time complexity is O(n2<sup>n</sup>). The depth of the tree is n and a temporary copy is being made at each node, so space complexity is O(n<sup>2</sup>).

## Solution 2 : Resursion without string copy
Here we use the same slate to avoid string copies. Since we are using the same slate to maintain its state correctly we need to push the character before the recursive call and pop that character after the recursive call.
```C++
void permutations(string &slate, string &s, int i, vector<string> &res)
{
    if(slate.size() == s.size())
    {
        res.push_back(slate);
        return;
    }
    
    if(isalpha(s[i]))
    {
        slate.push_back(static_cast<char>(toupper(s[i])));
        permutations(slate, s, i + 1, res);
        slate.erase(slate.end() - 1);
        slate.push_back(static_cast<char>(tolower(s[i])));
        permutations(slate, s, i + 1, res);
        slate.erase(slate.end() - 1);
    }
    else
    {
        slate.push_back(s[i]);
        permutations(slate, s, i + 1, res);
        slate.erase(slate.end() - 1);
    }
}
```

Time complexity is still O(n2<sup>n</sup>), even though we are not doing copy at each node, copy at the leaf nodes is needed. But space complexity is O(n) as we are not having a temporary copy at each node now.

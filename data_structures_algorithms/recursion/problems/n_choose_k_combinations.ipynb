Given two integers n and k, find all the possible unique combinations of k numbers in range 1 to n.  
Example :- Input : {"n": 5,"k": 2} -> Output : {{1, 2},{1, 3},{1, 4},{1, 5},{2, 3},{2, 4},{2, 5},{3, 4},{3, 5},{4, 5}}.

## Solution : Recursion
The problem can be viewed as recursion, at each index the number(index) can be included or excluded. Collect the vectors when the slate size is equal to k.
```C++
void combinations(vector<int> &slate, int n, int k, int i, vector<vector<int>> &res)
{
    if(slate.size() == k)
    {
        res.push_back(slate);
        return;
    }
    
    if(i == n + 1)
    {
        return;
    }
    
    combinations(slate, n, k, i+1, res);
    slate.push_back(i);
    combinations(slate, n, k, i+1, res);
    slate.erase(slate.end()  - 1);
}
```

Time complexity is O(n2n), as there are 2n nodes and a copy at the leaf nodes is needed. Space complexity is O(n) as it is a DFS tree.

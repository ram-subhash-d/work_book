Given a string that might contain duplicate characters, find all the possible distinct subsets of that string. All the subset strings should be individually sorted. 
Example :- Input : {"s": "aab"} -> Output : {"", "a", "aa", "aab", "ab", "b"}.

## Solution : Recursion
We can use the solution that we used for generate all subsets of a set. In addition we should not add duplicate entries to the result. Also since we want all the substrings to be sorted we have send a sorted string to the recursive function.
```C++
void distinct_subsets(string &slate, int cur_i, string &s, vector<string> &res)
{
    if(cur_i == s.size())
    {
        if(std::find(res.begin(), res.end(), slate) == res.end())
        {
            res.push_back(slate);
        }
        return;
    }
    
    distinct_subsets(slate, cur_i + 1, s, res);
    slate.push_back(s[cur_i]);
    distinct_subsets(slate, cur_i + 1, s, res);
    slate.erase(slate.end() - 1);
}
```

Given an array of numbers with possible duplicates, return all of its permutations in any order.  
Example :- Input : {"arr": {1, 2, 2}} -> Output : {{1, 2, 2},{2, 1, 2},{2, 2, 1}}.

## Solution 1 : using recursion
The solution should be same as permute an array of unique integers, except that we maintain a set to not call onto duplicate branches of the tree. This will eliminate all the duplicate permutations.
```C++
void permutations(vector<int> &arr, int cur_i, vector<vector<int>> &res)
{
    if(cur_i == (arr.size() -1))
    {
        res.push_back(arr);
        return;
    }
    
    std::unordered_set<int> used;
    for(int i = cur_i; i < arr.size(); i++)
    {
        if(used.count(arr[i]) < 1)
        {
            std::swap(arr[cur_i], arr[i]);
            permutations(arr, cur_i+1, res);
            std::swap(arr[cur_i], arr[i]);
            used.insert(arr[i]);
        }
    }
}
```

Time complexity is O(n * n!) and space complexity is O(n), same as for permute an array of unique integers.

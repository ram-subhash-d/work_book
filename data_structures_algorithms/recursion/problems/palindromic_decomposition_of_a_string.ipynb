Find all palindromic decompositions of a given string s. A palindromic decomposition of string is a decomposition of the string into substrings, such that all those substrings are valid palindromes.  
Example :- Input : {"s": "abracadabra"} -> Output : {"a|b|r|a|c|ada|b|r|a", "a|b|r|aca|d|a|b|r|a", "a|b|r|a|c|a|d|a|b|r|a"}.

## Solution : Recursion
This is like a substr problem, in every recursive find all the palindromes from current position to the end of the string.
```C++
bool is_substr_palindrome(string &s, int s_i, int e_i)
{
    string substr = s.substr(s_i, e_i - s_i + 1);
    string rsubstr(substr.rbegin(), substr.rend());
    
    return substr == rsubstr;
}

void palindromic_decompositions(std::string slate, string &s, int pos, vector<string> &res)
{
    if(pos == s.size())
    {
        slate.erase(slate.end() - 1);
        res.push_back(slate);
        return;
    }
    
    for(int i = pos; i < s.size(); i++)
    {
        if(is_substr_palindrome(s, pos, i))
        {
            palindromic_decompositions(slate + s.substr(pos, i - pos + 1) + "|", s, i + 1, res);
        }
    }
}
```

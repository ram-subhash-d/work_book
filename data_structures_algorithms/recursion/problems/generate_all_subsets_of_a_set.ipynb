Generate ALL possible subsets of a given set. The set is given in the form of a string s containing distinct lowercase characters 'a' - 'z'.  
Example :- Input : {"s": "xy"} -> Output : {"", "x", "y", "xy"}.

## Solution : Recursion
The problem can be viewed as recursion, at each index the character at that index can be included or excluded.
```C++
void subsets(string &slate, string &s, int i, vector<string> &res)
{
    if(i == s.size())
    {
        res.push_back(slate);
        return;
    }
    
    subsets(slate, s, i+1, res);
    slate = slate + s[i];
    subsets(slate, s, i+1, res);
    slate.erase(slate.end() - 1);
}
```

Time complexity is O(n2<sup>n</sup>), as there are 2<sup>n</sup> nodes and a copy at the leaf nodes is needed. Space complexity is O(n) as it is a DFS tree.

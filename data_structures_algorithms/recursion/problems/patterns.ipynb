Recursion is used when we want to solve exhaustive enumeration problems. Below are some of the the well known patterns to solve enumeration problems.
* Subsets : These are combination problems. View the problem as in each iteration a character has a possibility to be included or not included(excluded) in the subset. In general the recursive function should be passed the slate and the index for that recursion. All the subsets can be found at the leaf nodes, that is when i is equal to input size - 1. If there are any contraints in the problem, the conditions should be applied at the top of the recursive function, before the base(leaf) condition. This makes it a backtracking problem. It is called backtracking since some of the tree branches are not pursued till the leaf due to the constraints, thus improving the performance of the over all algorithm.
* Permutations : In permutations again repetitions can be allowed or not allowed. View these problems as in each iteration what all characters/numbers are possible at that index. If repetitions are not allowed, if a character is used in a recursive index, that character should not be allowed to be used in the tree below that node. So here along with the slate and index we should pass the remaining characters to the recursive function. All concepts related to backtracking apply to permutation problems too.

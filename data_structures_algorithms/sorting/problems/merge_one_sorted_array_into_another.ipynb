First array has n positive numbers, and they are sorted in the non-descending order. Second array has 2n numbers: first n are also positive and sorted in the same way but the last n are all zeroes.Merge the first array into the second and return the latter. You should get 2n positive integers sorted in the non-descending order.  
Example :- First : {1, 3, 5} Second : {2, 4, 6, 0, 0, 0} -> Output : {1, 2, 3, 4, 5, 6}.

## Solution : using 3 pointers
Use 3 pointers i, j and r, to point to nth element in the first array, nth element in the second array and last element(2n) in the first array respectively. Compare values at i and j and push them into r. 
```C++
vector<int> merge_one_into_another(vector<int> &first, vector<int> &second) 
{
    // Write your code here.
    
    //3 indexes
    int i = first.size() - 1;
    int j = i;
    int r = second.size() - 1;
    
    //push the elements into the second
    while(i >= 0)
    {
        if(j >= 0 && second[j] > first[i])
        {
            second[r] = second[j];
            j--;
        }
        else
        {
            second[r] = first[i];
            i--;
        }
        r--;
    }
    
    return {second};
}
```

There is only one pass through all the elements, so time complexity is O(n). No extra space is used so space complexity is constant O(1).

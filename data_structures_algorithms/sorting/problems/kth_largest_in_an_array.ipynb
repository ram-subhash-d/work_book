Given an array of integers, find the k-th largest number in it.
Example :- {"numbers": {5, 1, 10, 3, 2},"k": 2} -> Output : 5.

## Solution 1 : Sorting
```C++
int kth_largest_in_an_array(vector<int> &numbers, int k)
{   
    //Sort the elements in a decreasing order
    std::sort(numbers.begin(), numbers.end(), greater<int>());
    
    return numbers[k-1];
}
```

Since we are using sort time complexity is O(n * logn). No extra space used.

## Solution 1 : Priority queue(min heap)
```C++
int kth_largest_in_an_array(vector<int> &numbers, int k) 
{   
    //Maintain a prioprity queue with the k largest element at the top
    priority_queue<int, vector<int>, greater<int>> pq;
    
    //Go through the elements and push them
    for(int &n : numbers)
    {
        if(pq.size() < k)
        {
            pq.push(n);
        }
        else if(n > pq.top())
        {
            pq.pop();
            pq.push(n);
        }
    }
    
    return pq.top();
}
```

Since we are using a priority queue of size k over n elements, time complexity is O(n * logk). Space complexity is O(k) for the priority queue.

Given an initial list along with another list of numbers to be appended with the initial list and an integer k, return an array consisting of the k-th largest element after adding each element from the first list to the second list.  
Example :- {"k": 2,"initial_stream": {4, 6},"append_stream": {5, 2, 20}} -> Output : {5, 5, 6}.

## Solution : using priority queue(min heap)
```C++
vector<int> kth_largest(int k, vector<int> &initial_stream, vector<int> &append_stream) 
{    
    //Build the priority queue that maintains the top k elements
    priority_queue<int, vector<int>, greater<int>> pq;
    
    //initial list
    for(int &ele : initial_stream)
    {
        if(pq.size() < k)
        {
            pq.push(ele);
        }
        else if(ele > pq.top())
        {
            pq.pop();
            pq.push(ele);
        }
    }
    
    //Result
    vector<int> res;
    
    //build res as append stream is processed
    for(int &ele : append_stream)
    {
        if(pq.size() < k)
        {
            pq.push(ele);
        }
        else if(ele > pq.top())
        {
            pq.pop();
            pq.push(ele);
        }
        
        res.push_back(pq.top());
    }
    
    return {res};
}
```

Since we are maintaining a priority queue of size k going over m + n elements, time complexity is O((m + n)logk). Space complexity is O(k) used for the priority queue.

Given coordinates of a point p and n other points on a two-dimensional surface, find k points out of n which are the nearest to point p. Distance is measured by the standard euclidean method.  
Example :- Input: {"p_x": 1,"p_y": 1,"k": 1,"n_points": {{0, 0},{1, 0}}} -> Output : {{1, 0}}.

## Solution 1 : Priority Queue
Use a priority queue to save the indexes of k least distance points.
```C++
vector<vector<int>> nearest_neighbours(int p_x, int p_y, int k, vector<vector<int>> &n_points) {
    // Write your code here.
    std::priority_queue<pair<double, int>, vector<pair<double, int>>, greater<pair<double, int>>> pq;
    
    //Push into the priority queue, based on the distance
    double d = 0;
    for(int i = 0; i < n_points.size(); i++)
    {
        d = std::hypot(n_points[i][0] - p_x, n_points[i][1] - p_y);
        pq.push({d, i});
    }
    
    //Add the top k elements to the response
    vector<vector<int>> res;
    while(k > 0)
    {
        pair<double, int> v = pq.top();
        res.push_back(n_points[v.second]);
        
        pq.pop();
        k--;
    }
    
    return {res};
}
```

As we compute the distance of each point present in array and also maintain a max heap of size k, the time complexity is O(n * log(k)). Space complexity is O(k) for the heap.

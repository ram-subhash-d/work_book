Given a list of characters, sort it in the non-decreasing order based on ASCII values of characters.  
Example :- Input: {"a", "s", "d", "f", "g", "*", "&", "!", "z", "y"} -> Output : {"!", "&", "*", "a", "d", "f", "g", "s", "y", "z"}.

## Solution 1 : Sorting
Sort the array and return it.
```C++
vector<char> sort_array(vector<char> &arr) 
{
    // Write your code here.
    std::sort(arr.begin(), arr.end());
    return {arr};
}
```

Since we are using the built-in sort function, which uses randomized quick sort to sort an array, average case time complexity is O(n * log(n)). No extra space so space complexity is O(1).

## Solution 2: Counting
Here we use counting sort, where we use an int array of size 128 to store the counts of ascii values of the input characters, then we fill back the input array from the counting array. 
```C++
vector<char> sort_array(vector<char> &arr) 
{
    // Write your code here.
    std::array<int, 128> ascii_count{0};
    
    //fill up the ascii count
    for(char c : arr)
    {
        ascii_count[(int)c]++;
    }
    
    //fillup the output array from the counts array
    int j = 0;
    for(int i = 0; i < ascii_count.size(); i++)
    {
        while(ascii_count[i] > 0)
        {
            arr[j] = char(i);
            j++;
            ascii_count[i]--;
        }
    }
    
    
    return {arr};
}
```

Time complexity is O(n) as we loop through the characters only once. Since the counting array is only of size 128, no extra space is used, so space complexity is O(1).

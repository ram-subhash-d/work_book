Given an integer array and a number k, find the k most frequent elements in the array.  
Example :- {"arr": {1, 2, 3, 2, 4, 3, 1},"k": 2} -> Output : {3, 1}.

## Solution 1 : using a hash map
Use a hash map to count the frequency of the numbers, then use a priority queue to get the top k frequent elements.
```C++
vector<int> find_top_k_frequent_elements(vector<int> &arr, int k) 
{
    // Write your code here.
    
    //hashmap to store the integer and its frequency
    std::unordered_map<int, int> frequency_map;
    
    //Fill up the frequency map
    for(int &ele : arr)
    {
        if(frequency_map.count(ele) > 0)
        {
            frequency_map[ele]++;
        }
        else
        {
            frequency_map[ele] = 1;
        }
    }
    
    priority_queue<pair<int, int>, vector<pair<int, int>>, greater<pair<int, int>>> count_pq;
    for(pair<int, int> ele : frequency_map)
    {
        if(count_pq.size() < k)
        {
            count_pq.push({ele.second, ele.first});
        }
        else if((count_pq.size() == k)  && (count_pq.top().first < ele.second))
        {
            count_pq.pop();
            count_pq.push({ele.second, ele.first});
        }
    }
    
    //Make the result to be returned
    vector<int> res;
    while(!count_pq.empty())
    {
        res.push_back(count_pq.top().second);
        count_pq.pop();
    }
    
    return {res};
}
```

Here the second loop is the most time comsuming loop, outer loop is of order n and priority queue of size k take log k time, so time complexity is O(n * logk). We have a extra map of size n and priority queue of size k, so space complexity is O(n+k).

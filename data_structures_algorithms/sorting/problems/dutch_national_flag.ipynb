Given some balls of three colors arranged in a line, rearrange them such that all the red balls go first, then green and then blue ones. Do rearrange the balls in place. A solution that simply counts colors and overwrites the array is not the one we are looking for.  
Example :- Input : {"G", "B", "G", "G", "R", "B", "R", "G"} -> Output : {"R", "R", "G", "G", "G", "G", "B", "B"}. Do this in one pass over the array, not two passes.

```C++
vector<char> dutch_flag_sort(vector<char> &balls) {
    // Write your code here.
    
    //Start indexes
    int r_idx = 0;
    int b_idx = balls.size() - 1;
    int curr_idx = 0;
    
    //If red swap with the first green
    //If green leave it in place
    //If blue swap it with index before blue's, don't increament current index
    //as any color could come at this location
    while(curr_idx <= b_idx)
    {
        if(balls[curr_idx] == 'R')
        {
            std::swap(balls[r_idx++], balls[curr_idx++]);
        }
        else if(balls[curr_idx] == 'G')
        {
            curr_idx++;
        }
        else
        {
            std::swap(balls[b_idx--], balls[curr_idx]);
        }
    }
    
    
    return {balls};
}
```

Since one loop is used time complexity is linear O(n). No extra space is used so space complexity is constant O(1).

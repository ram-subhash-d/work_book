Given an array of numbers, rearrange them in-place so that even numbers appear before odd ones.  
Example :- Input : {1, 2, 3, 4} -> Output : {4, 2, 3, 1}. The order within the group of even numbers does not matter, same with odd numbers.

```C++
vector<int> segregate_evens_and_odds(vector<int> &numbers) {
    // Write your code here.
    int i = 0;
    int j = numbers.size() - 1;
    
    //Move the 2 indexes from the 2 ends
    //If even and odd are not in the correct half swap
    while(i < j)
    {
        if(numbers[i]%2 == 0)
        {
            i++;
            continue;
        }
        else if(numbers[j]%2 == 1)
        {
            j--;
            continue;
        }
        else
        {
            std::swap(numbers[i], numbers[j]);
            i++;
            j--;
        }
    }
    
    
    return {numbers};
}
```

Since one loop is used time complexity is linear O(n). No extra space is used so space complexity is constant O(1).
